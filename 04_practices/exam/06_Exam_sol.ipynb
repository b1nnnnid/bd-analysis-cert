```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/BigData/Exam/6/2_1_bp_train.csv')
test = pd.read_csv('/content/drive/MyDrive/BigData/Exam/6/2_1_bp_test.csv')

# 답안 제출 참고
# 아래 코드는 예시이며 변수명 등 개인별로 변경하여 활용
# pd.DataFrame변수.to_csv('result.csv', index=False)

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/BigData/Exam/6/2_1_bp_train.csv')
test = pd.read_csv('/content/drive/MyDrive/BigData/Exam/6/2_1_bp_test.csv')

# X, Y 데이터 셋 분리
X = train.drop(['BP'], axis=1)
y = train['BP']

X_full = pd.concat([X, test], axis=0)
#print(X_full.shape)

# 범주형 변수 인코딩 - One-Hot Encoding
X_full = pd.get_dummies(X_full)
#print(X_full)

# 학습, 검증 데이터 셋 분할
X_train = X_full[:train.shape[0]]
X_test = X_full[train.shape[0]:]
#print(X_train.shape, X_test.shape)

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, stratify=y)
#print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)

# Randomforest 활용 모델 학습
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

# f1_score, Accuracy 활용 평가
from sklearn.metrics import f1_score, accuracy_score
y_pred = model.predict(X_val)
f1 = f1_score(y_val, y_pred, average='macro')
acc = accuracy_score(y_val, y_pred)
print(f1, acc)

# test 데이터 예측 및 결과 저장 (pred 1개 컬럼)
y_pred = model.predict(X_test)
result = pd.DataFrame(y_pred, columns=['pred'])
result.to_csv('result.csv', index=False)
```

2. 감기약의 부작용 비율과 항암제의 부작용 관찰 값이 통계적으로 유의미하게 차이가 있는지 확인하기 위한 카이제곱 검정 통계량을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd
from scipy.stats import chisquare

df = pd.read_csv('/content/drive/MyDrive/BigData/Exam/6/3_1_side_effects.csv')

# -------------------------------
# (1) 무증상 비율 계산
# -------------------------------

# '무증상' 항목의 개수를 센다
no_counts = sum(df['부작용'] == '무증상')

# 전체 환자 수 (총 데이터 행 수) 계산
total_counts = len(df)

# 무증상 비율 = 무증상 개수 / 전체 환자 수
no_symptom_ratio = round(no_counts / total_counts, 3)

# 결과 출력
print("① 무증상 비율:", no_symptom_ratio)

# -------------------------------
# (2) 카이제곱 통계량 계산
# -------------------------------

# 부작용 항목 리스트 (명시적 순서 설정)
# 이 순서에 따라 관측값과 기대값의 인덱스가 정확히 일치해야 함
# ⚠️ 인덱스 순서가 다르면 통계 검정 결과가 왜곡될 수 있음
side_effects = ['무증상', '속쓰림', '조금아픔', '아픔']

# 실제 관찰된 부작용 빈도수를 계산
# value_counts(): 각 부작용별 개수 계산
# reindex(): side_effects 순서로 정렬하며, 없는 항목은 fill_value=0으로 채움 (예외 방지!)
observed_counts = df['부작용'].value_counts().reindex(side_effects, fill_value=0).values

# 문제에서 주어진 감기약 부작용 비율 (기대 비율)
expected_ratios = [0.6, 0.25, 0.1, 0.05]

# 기대 빈도 = 기대 비율 * 전체 샘플 수
# 카이제곱 검정은 "기대 빈도 vs 실제 관찰값"을 비교하여 유의미한 차이가 있는지 판단
expected_counts = [expected_ratios[0] * total_counts,
                   expected_ratios[1] * total_counts,
                   expected_ratios[2] * total_counts,
                   expected_ratios[3] * total_counts]
# expected_counts = [r * total_counts for r in expected_ratios]
# ↑ 위 코드를 리스트 컴프리헨션을 활용하여 더 간결하게 표현한 방법
#   기능은 동일하지만 코드가 짧고 깔끔해지므로 참고

# -------------------------------
# 카이제곱 검정 수행
# -------------------------------
# chisquare(f_obs, f_exp)
# - f_obs: 실제 관찰된 빈도 (observed_counts)
# - f_exp: 기대되는 빈도 (expected_counts)
# 두 리스트는 같은 순서, 같은 길이여야 하며, 각 항목이 서로 1:1 대응돼야 함
chi_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# -------------------------------
# (2) 결과 출력 - 카이제곱 통계량
# -------------------------------
print("② 카이제곱 통계량:", round(chi_stat, 3))

# -------------------------------
# (3) 결과 출력 - p-value
# -------------------------------
print("③ p-value:", round(p_value, 3))
```

주어진 데이터들을 기반으로 다중 선형 회귀 분석을 수행하고, 아래 소문항에 답하시오. (단, 모델은 절편을 포함한다.)

[제공 데이터]

2. RestingBP : 안정 시 혈압 (mmHg)
3. Cholesterol : 콜레스테롤 수치 (mg/dL)

[문제]

 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①에서 구축한 모델에서 Age에 대한 p-value를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ①에서 구축한 모델을 이용하여, 아래 새로운 데이터 정보를 입력했을 때 예상 BPM을 구하여라.
(Age=45, RestingBP=120, Cholesterol=190)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성


```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/Exam/6/3_2_health.csv')

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd
import statsmodels.formula.api as smf

df = pd.read_csv('/content/drive/MyDrive/BigData/Exam/6/3_2_health.csv')

# 다중 선형회귀 분석 수행
model = smf.ols('BPM ~ Age + RestingBP + Cholesterol', data=df).fit()

# ① Cholesterol의 회귀 계수
chol_coef = round(model.params['Cholesterol'], 3)
print("① Cholesterol의 회귀 계수:", chol_coef)

# ② Age의 p-value
age_pval = round(model.pvalues['Age'], 3)
print("② Age의 p-value:", age_pval)

# ③ 새로운 데이터 입력 시 예측값 계산
new_data = pd.DataFrame({'Age': [45], 'RestingBP': [120], 'Cholesterol': [190]})
predicted_bpm = round(model.predict(new_data)[0], 3)
print("③ 예측 BPM:", predicted_bpm)
```
