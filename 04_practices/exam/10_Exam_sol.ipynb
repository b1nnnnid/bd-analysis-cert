```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/Exam/10/1_1_math.csv')

# 서브주제별 정답 수 및 전체 응답 수 계산
subtopic_accuracy = df.groupby('Subtopic')['Answer'].agg(['sum', 'count'])

# 정답률 계산
subtopic_accuracy['Accuracy'] = subtopic_accuracy['sum'] / subtopic_accuracy['count']

# 정답률 내림차순 정렬 후 세 번째 값 추출
third_highest = subtopic_accuracy['Accuracy'].sort_values(ascending=False)
print("①", round(third_highest.iloc[2], 3))
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/Exam/10/1_2_sales.csv')

# 날짜에서 연도와 월 추출
df['order_date'] = pd.to_datetime(df['order_date'])
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month

# 1. 연도/월별 총매출 계산 및 정렬
monthly_sales = df.groupby(['year', 'month'])['price'].sum().sort_values(ascending=False)

# 2. 두 번째로 높은 매출액
second_highest_sales = int(monthly_sales.iloc[1])

# 3. 네 번째로 높은 연/월 추출 후 카테고리별 매출 계산
fourth_year, fourth_month = monthly_sales.index[3]
filtered_df = df[(df['year'] == fourth_year) & (df['month'] == fourth_month)]
category_sales = filtered_df.groupby('category')['price'].sum().sort_values(ascending=False)
highest_category_sales = int(category_sales.iloc[0])

# 출력
print("①", second_highest_sales)
print("②", highest_category_sales)
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/10/1_3_spam.csv")

# 1. 단어 수를 세는 함수 정의
def count_words(text):
    return len(text.split())

# 2. apply에 함수 적용해서 word_count 칼럼 생성
df['count'] = df['Message'].apply(count_words)
df['count'] = df['Message'].str.split().str.len()

# 3. Category(spam/ham)별 평균 단어 수 계산
avg_word_count = df.groupby('Category')['count'].mean()

# 4. 평균 차이 절대값 계산 및 반올림
result = round(abs(avg_word_count['spam'] - avg_word_count['ham']), 3)

print("①", result)
```

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/BigData/Exam/10/2_1_gas_train.csv')
test = pd.read_csv('/content/drive/MyDrive/BigData/Exam/10/2_1_gas_test.csv')

# 답안 제출 참고
# 아래 코드는 예시이며 변수명 등 개인별로 변경하여 활용
# pd.DataFrame변수.to_csv('result.csv', index=False)

# 여기에 코드를 작성하시오.
```

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/BigData/Exam/10/2_1_gas_train.csv')
test = pd.read_csv('/content/drive/MyDrive/BigData/Exam/10/2_1_gas_test.csv')

# X, y 분리
X = train.drop(['GAS_TOTL'], axis=1)
y = train['GAS_TOTL']

# 학습 + 테스트 데이터 통합
X_full = pd.concat([X, test], axis=0)

# 범주형 변수 One-Hot 인코딩
X_full = pd.get_dummies(X_full)

# 학습, 테스트 셋 재분리
X_train = X_full[:len(train)]
X_test = X_full[len(train):]

# 학습/검증 셋 분리
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, random_state=42)

# 모델 학습 - RandomForest
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# 성능 평가
from sklearn.metrics import root_mean_squared_error, r2_score
y_val_pred = model.predict(X_val)
rmse = root_mean_squared_error(y_val, y_val_pred)
r2 = r2_score(y_val, y_val_pred)

print("RMSE:", rmse)
print("R2 Score:", r2)

# 최종 예측 및 제출파일 생성
y_test_pred = model.predict(X_test)
submission = pd.DataFrame({'pred': y_test_pred})
submission.to_csv('result.csv', index=False)
```

[제공 데이터]
1. Age : 직원 나이
2. MonthlyIncome : 월급
3. OverTime : 야근 여부 (0: 없음, 1: 있음)
4. Attrition : 퇴사 여부 (0: 재직, 1: 퇴사)

[문제]
1. Age, MonthlyIncome, OverTime을 독립변수로 사용하여 로지스틱 회귀분석을 수행하고, 가장 통계적으로 유의한 변수에 대하여 회귀계수를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①에서 구축한 모형을 이용하여, 나이가 1살 증가할 때마다 퇴사할 오즈비(Odds Ratio)를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ①에서 구축한 모형을 이용하여, 아래 새로운 직원 정보가 주어졌을 때, 퇴사할 확률을 구하여라.
  (Age=45, MonthlyIncome=4500, OverTime=1)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/10/3_1_employee.csv")

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/10/3_1_employee.csv")

import statsmodels.formula.api as smf
import numpy as np

# 1. 로지스틱 회귀 분석 수행
model = smf.logit('Attrition ~ Age + MonthlyIncome + OverTime', data=df).fit()

print(model.summary())

# ① 월급의 회귀계수 출력
monthly_income_coef = round(model.params['MonthlyIncome'], 3)
print("① MonthlyIncome 회귀계수:", monthly_income_coef)

# ② 나이의 오즈비 계산 (exp(계수))
age_odds_ratio = round(np.exp(model.params['Age']), 3)
print("② 나이 오즈비:", age_odds_ratio)

# ③ 예측 수행
new_employee = pd.DataFrame({
    'Age': [45],
    'MonthlyIncome': [4500],
    'OverTime': [1]
})

predicted = model.predict(new_employee)
predicted_prob = round(predicted.iloc[0], 3)
print("③ 퇴사 확률 예측:", predicted_prob)
```
