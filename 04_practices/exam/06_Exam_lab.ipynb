1-1) 제공된 데이터 fire_truck.csv는 소방서 구급차의 출동 및 도착 시간을 포함한 로그 데이터이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 [제출 형식]에 맞춰 입력하시오.

[수행 순서]
1. 구급차 ID별로 도착시간과 출동시간의 차이를 계산하여 이동시간을 구한다.
 ※ 이동시간은 '분' 단위로 변환한다.
2. 구급차 ID별 평균 이동시간을 계산한다.
3. 평균 이동시간이 가장 오래 걸리는 구급차의 이동시간을 입력한다.

[제출 형식]
반올림하여 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv('../data/6/1_1_fire_truck.csv')

# 여기에 코드를 작성하시오.

# print(df.info())
# print(df.head(10))

df['출동시간'] = pd.to_datetime(df['출동시간']) # text 먼저 datetime으로 변환
df['도착시간'] = pd.to_datetime(df['도착시간'])

df['이동시간'] = (df['도착시간']-df['출동시간']).dt.total_seconds()/60 # dt.total_seconds() 까먹지 말기
time_id_avg = df.groupby('id')['이동시간'].mean()
# print(time_id_avg)

ans = round(time_id_avg.max()) # 반올림은 int말고 round
print("ans: ", ans)
```

# 제2유형

제공된 학습용 데이터(bp_train.csv)는 환자의 정보(나이, 성별, 키, 몸무게, BMI, 콜레스테롤, 흡연 유무 등)를 포함하고 있다. 이를 바탕으로 혈압 상태를 예측하는 모델을 개발하고, 가장 우수한 모델을 평가 데이터(bp_test.csv)에 적용하여 예측 결과를 도출하시오. 예측 결과는 아래의 [제출 형식]을 준수하여, CSV 파일로 생성하는 코드를 제출하시오.
 * 예측결과는 F1 Score (macro) 평가지표에 따라 평가함


[제공 데이터]
1. bp_train.csv : 학습용 데이터 (400 건)
2. bp_test.csv : 평가용 데이터 (200 건, 혈압 라벨 없음)

[데이터 컬럼]
1. Age : 나이
2. Sex : 성별 (F: 여성, M: 남성)
3. Height_cm : 키 (cm)
4. Weight_kg : 몸무게 (kg)
5. BMI : 체질량지수
6. Cholesterol : 콜레스테롤 수치 (HIGH, NORMAL, LOW)
7. Smoker : 흡연 여부 (YES, NO)
8. BP : 혈압 상태 (HIGH, NORMAL, LOW)

[제출 형식]
1. 제출 파일명 : result.csv
2. 제출 컬럼명 : pred
3. 예측 결과 개수 : 200

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('../data/6/2_1_bp_train.csv')
test = pd.read_csv('../data/6/2_1_bp_test.csv')

# 답안 제출 참고
# 아래 코드는 예시이며 변수명 등 개인별로 변경하여 활용
# pd.DataFrame변수.to_csv('result.csv', index=False)

# 여기에 코드를 작성하시오.

# print(train.info())
# print(test.info())

# print(train.isnull().sum())
# print(test.isnull().sum()) 

X = train.drop('BP', axis=1)
y = train['BP']

X_full = pd.concat([X, test], axis=0)
# print(X_full.shape)

X_full = pd.get_dummies(X_full)
# print(X_full)

X_train = X_full[:train.shape[0]]
test = X_full[train.shape[0]:]

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, stratify=y)
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, max_depth=8)
model.fit(X_train, y_train)

# from lightgbm import LGBMClassifier
# model = LGBMClassifier()
# model.fit(X_train, y_train)

# %pip install xgboost
# from xgboost import XGBClassifier
# model = XGBClassifier()
# model.fit(X_train, y_train)

from sklearn.metrics import accuracy_score, f1_score
y_pred = model.predict(X_val)
acc = accuracy_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred, average='macro')
print(acc, f1)

y_pred = model.predict(test)
result = pd.DataFrame(y_pred, columns=['pred'])
print(result)
result.to_csv('result_6.csv', index=False)
```

1. 귀무가설(H0): 항암제 부작용 분포는 감기약 부작용 분포와 차이가 없다.
2. 대립가설(H1): 항암제 부작용 분포는 감기약 부작용 분포와 차이가 있다.


[제공 데이터]

 ※ 부작용 항목: 무증상, 속쓰림, 조금아픔, 아픔
2. 감기약 부작용 분포(비율)
 ※ 무증상(60%), 속쓰림(25%), 조금아픔(10%), 아픔(5%)

[문제]
1. 항암제 투여 환자 중 무증상 비율을 0과 1 사이의 값으로 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. 감기약의 부작용 비율과 항암제의 부작용 관찰 값이 통계적으로 유의미하게 차이가 있는지 확인하기 위한 카이제곱 검정 통계량을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②에서 수행한 검정의 p-value 값을 구하여라.
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv('../data/6/3_1_side_effects.csv')

# 여기에 코드를 작성하시오.
# print(df.value_counts()) / no_counts = sum(df['부작용']=='무증상')
# print(len(df))
ans1 = round(55/100,3)
print("ans1: ", ans1)

from scipy import stats
exp = [60, 25, 10, 5] # 기대빈도 설정(감기약 부작용 분포)
obs = [55, 35, 4, 6] # df.value_counts.values 했다가 값끼리 순서 다를 수도... 안전하게 직접 적자
t1, p1 = stats.chisquare(obs, exp)
ans2 = round(t1,3)
print("ans2: ", ans2)
ans3 = round(p1,3)
print("ans3: ", ans3)
```

3-2) 제공된 데이터(health.csv)는 150명의 사람에 대해 건강 관련 정보와 심박수(BPM)를 기록한 자료이다.
주어진 데이터들을 기반으로 다중 선형 회귀 분석을 수행하고, 아래 소문항에 답하시오. (단, 모델은 절편을 포함한다.)

[제공 데이터]
1. Age : 나이 (세)
2. RestingBP : 안정 시 혈압 (mmHg)
3. Cholesterol : 콜레스테롤 수치 (mg/dL)
4. BPM : 1분당 심박수

[문제]
1. Age, RestingBP, Cholesterol을 독립변수, BPM을 종속변수로 하여 다중선형회귀분석을 수행하고, Cholesterol의 회귀 계수 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①에서 구축한 모델에서 Age에 대한 p-value를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ①에서 구축한 모델을 이용하여, 아래 새로운 데이터 정보를 입력했을 때 예상 BPM을 구하여라.
(Age=45, RestingBP=120, Cholesterol=190)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성


```python
import pandas as pd

df = pd.read_csv('../data/6/3_2_health.csv')

# 여기에 코드를 작성하시오.
import statsmodels.formula.api as smf

model = smf.ols('BPM ~ Cholesterol + Age + RestingBP', data= df).fit() # olf 아니고 ols, 끝에 fit 빼먹지 말기
print(model.summary())
ans1 = round(-0.1876,3) # model.params['Cholesterol']
print("ans1 :",ans1)
ans2 = 0.408 # model.pvalues['Age']
print("ans2: ", ans2)

new_data = pd.DataFrame({'Age':[45], 'RestingBP':[120], "Cholesterol": [190]})
pred = model.predict(new_data)
ans3 = round(pred[0],3)
print("ans3: ",ans3)
```
