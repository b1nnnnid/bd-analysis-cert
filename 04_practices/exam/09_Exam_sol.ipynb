4. 만족도 : 1~10 사이의 만족도 점수

[수행 순서]
1. 만족도의 결측치는 전체 평균값으로 대체한다.
2. 근속연수의 결측치는 동일한 부서 내에서 같은 성과등급을 가진 직원들의 평균 근속연수로 대체한다. 이때, 소수점 이하는 버리고 정수로 대체한다.
3. 부서 평균 만족도가 가장 높은 부서를 찾고, 해당 부서 내 성과등급별 평균 만족도를 계산한다. 이 중 가장 낮은 성과등급의 평균 근속연수를 A라고 한다.
4. 각 지원의 ‘연봉만족도 = 근속연수 / 만족도’ 값을 계산하고, 이 값이
두 번째로 큰 직원의 교육참가횟수를 B라고 한다.

[문제]
1. 수행 순서 3)과 4)에서 구한 A, B의 값의 합(A+B)을 입력하시오.
 ☞ [제출 형식] 버림하여 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/9/1_3_hr.csv")

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/9/1_3_hr.csv")

# 1. 만족도 결측치 → 전체 평균으로 대체
df["만족도"] = df["만족도"].fillna(df["만족도"].mean())

# 2. 근속연수 결측치 → 부서 + 성과등급별 평균으로 대체 (소수점 버림)
df["근속연수"] = df["근속연수"].fillna(
    df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
)
df["근속연수"] = np.floor(df["근속연수"]).astype(int)

# 3. 부서별 만족도 평균 → 가장 높은 부서 선택
top_dept = df.groupby("부서")["만족도"].mean().idxmax()

# 해당 부서 내 성과등급별 만족도 평균 → 가장 낮은 등급
lowest_grade = df[df["부서"] == top_dept].groupby("성과등급")["만족도"].mean().idxmin()

# 해당 성과등급의 평균 근속연수
part3_value = df[(df["부서"] == top_dept) & (df["성과등급"] == lowest_grade)]["근속연수"].mean()

# 4. 연봉_만족도 = 근속연수 ÷ 만족도
df["연봉_만족도"] = df["근속연수"] / df["만족도"]

# 두 번째로 큰 직원의 교육참가횟수
part4_value = df.nlargest(2, "연봉_만족도").iloc[-1]["교육참가횟수"]

# 5 결과
result = int(part3_value + part4_value)
print(result)
```

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/BigData/Exam/9/2_1_pesticide_train.csv')
test = pd.read_csv('/content/drive/MyDrive/BigData/Exam/9/2_1_pesticide_test.csv')

# X, Y 데이터 셋 분리
X = train.drop(['Pesticide_Result'], axis=1)
y = train['Pesticide_Result']

X_full = pd.concat([X, test], axis=0)
#print(X_full.shape)

# 범주형 변수 인코딩 - One-Hot Encoding
X_full = pd.get_dummies(X_full)
#print(X_full)

# 학습, 검증 데이터 셋 분할
X_train = X_full[:train.shape[0]]
X_test = X_full[train.shape[0]:]
#print(X_train.shape, X_test.shape)

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, stratify=y)
#print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)

# Randomforest 활용 모델 학습
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

# f1_score, Accuracy 활용 평가
from sklearn.metrics import f1_score, accuracy_score
y_pred = model.predict(X_val)
f1 = f1_score(y_val, y_pred, average='macro')
acc = accuracy_score(y_val, y_pred)
print(f1, acc)

# test 데이터 예측 및 결과 저장 (pred 1개 컬럼)
y_pred = model.predict(X_test)
result = pd.DataFrame(y_pred, columns=['pred'])
result.to_csv('result.csv', index=False)
```

[제공 데이터]

1. f1, f2, f3, f4, f5를 독립변수로 하여 학습용(train) 데이터를 이용해 다중선형회귀분석을 수행하고, 절편을 제외한 독립변수 중 p값이 0.05보다 작은 변수의 개수를 구하여라.
 ☞ [제출 형식] 정수(integer)로 작성
2. ①의 결과를 바탕으로 유의수준 0.05 기준에서 유의하지 않은 독립변수를 제거한 후, 남은 변수만을 이용하여 다시 다중선형회귀분석을 수행한다. 이 모델로 train 데이터의 design 예측값을 산출한 뒤, 예측값과 실제 design값 사이의 피어슨 상관계수를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성  
3. ②에서 구축한 모형을 시험용(test) 데이터에 적용하여 예측값을 구하고, RMSE를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/9/3_1_design.csv")

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/9/3_1_design.csv")

import statsmodels.formula.api as smf
from sklearn.metrics import root_mean_squared_error

train, test = df.iloc[:140].copy(), df.iloc[140:].copy()

# ① 모든 변수 포함한 다중회귀모형 적합
model_full = smf.ols("design ~ f1 + f2 + f3 + f4 + f5", data=train).fit()
print(model_full.summary())

pvals = model_full.pvalues.drop("Intercept")
sig_vars = list(pvals[pvals < 0.05].index)
ans1 = len(sig_vars)
print("① 유의한 변수 개수:", ans1)

# ② 유의하지 않은 f1, f3, f5 제거 후, 예측값과 실제값의 피어슨 상관계수 구하기
model_sig = smf.ols('design ~ f2 + f4', data=train).fit()
train["pred"] = model_sig.predict(train)
ans2 = round(train[["design", "pred"]].corr().iloc[0, 1], 3)
print("② 예측값과 실제값의 피어슨 상관계수:", ans2)

# ③ test 데이터 예측 및 RMSE 계산
test["pred"] = model_sig.predict(test)
ans3 = round(root_mean_squared_error(test["design"], test["pred"]),3)
print("③ RMSE 계산:", ans3)
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/9/3_2_retention.csv")

import statsmodels.formula.api as smf
import numpy as np

# 1. 로지스틱 회귀 분석 수행
model = smf.logit('Leave ~ MonthlyFee + ServiceMonths + PhonePlan + DeviceProtect', data=df).fit()
print(model.summary())

# ①  MonthlyFee의 p-value
p_monthly = round(model.pvalues["ServiceMonths"], 3)
print("① ServiceMonths p값:", p_monthly)

# ② PhonePlan의 오즈비
odds_phone = round(np.exp(model.params["PhonePlan"]), 3)
print("② PhonePlan 오즈비:", odds_phone)

# ③ 예측 확률이 0.3 초과인 고객 수
pred_prob = model.predict(df)
count_over_03 = int((pred_prob > 0.3).sum())
print("③ 예측확률 0.3초과 고객 수:", count_over_03)
```
