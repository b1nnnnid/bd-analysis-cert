3. 전체 기간(2001~2021)의 배출량 연평균 성장률(CAGR)을 계산한다.
※ 𝑪𝑨𝑮𝑹=((𝟐𝟎𝟐𝟏년도 총 배출량)/(𝟐𝟎𝟎𝟏년도 총 배출량))^((𝟏/(𝟐𝟎𝟐𝟏−𝟐𝟎𝟎𝟏)))−𝟏

```python
import pandas as pd

df = pd.read_csv('../data/11/1_1_emission.csv')

# 1. 연도별 온실가스 배출량 1위 국가 계산
idx = df.groupby("year")["emission"].idxmax()
top_by_year = df.loc[idx, ["year", "country", "emission"]]

# 2. 연도별 1위 국가들 중, 가장 많이 1위 한 국가의 횟수 계산
country_counts = top_by_year["country"].value_counts(ascending=False).iloc[0]

# 3. 전체 기간의 CAGR 계산
total_by_year = df.groupby("year")["emission"].sum()

start_year, end_year = 2001, 2021

start_val = total_by_year.loc[start_year]
end_val   = total_by_year.loc[end_year]
cagr = (end_val / start_val) ** (1 / (start_year - end_year)) - 1

# 출력
print("①", country_counts)
print("②", round(cagr, 3))
```

2. 취소컬럼이 T인 행의 거래금액은 음수 기호(-)를 붙인다.

4. 상품코드별로 거래금액을 합산하여 총 거래금액이 가장 큰 상품코드의 총 거래금액을 구한다.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/BigData/Exam/11/1_3_marketing.csv')
# 경로는 본인 환경에 맞게 수정

# 1. cancel 컬럼 생성 (T/F)
df["cancel"] = np.where(df["transaction_code"].str[0] == "C", "T", "F")

# 2. cancel이 T인 amount를 음수로 변경
df["amount"] = np.where(df["cancel"] == "T", df["amount"] * -1, df["amount"])

# 3. 총 취소금액 절대값이 가장 큰 고객 → 그 고객의 취소 거래 중 |amount| 최대 거래코드
cancel_only = df[df["cancel"] == "T"]
cancel_sum = cancel_only.groupby("customer_id")["amount"].sum()
top_customer = cancel_sum.abs().idxmax()

tmp = cancel_only[cancel_only["customer_id"] == top_customer]
answer_1 = tmp.loc[tmp["amount"].abs().idxmax(), "transaction_code"]

# 4. 상품코드별 총 거래금액 합계 중 최대값
prod_sum = df.groupby("product_code")["amount"].sum()
answer_2 = int(prod_sum.max())

print("①", answer_1)
print("②", answer_2)
```

[제공 데이터]

 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

 ☞ [제출 형식] 정수(integer)로 작성
3. ①에서 구축한 모델에 대해 임계값(Threshold)=0.5로 분류했을 때의 민감도(Sensitivity) 를 구하여라. (Outcome=1을 양성으로 간주)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/11/3_1_diabetes.csv")

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.metrics import confusion_matrix

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/11/3_1_diabetes.csv")

# ① 로지스틱 회귀: Outcome ~ Age + Diabetes_History (No 기준)
model = smf.logit("Outcome ~ Age + Diabetes_History", data=df).fit()

# Diabetes_History[T.Pre]의 오즈비 → "No 대비 Pre 상태일 때 당뇨 발생 오즈 변화"
odds_ratio = np.exp(model.params["Diabetes_History[T.Pre]"])
print("①", round(odds_ratio, 3))

# ---------------------------------------------------------
# 참고: statsmodels에서는 C()를 이용해 범주형 변수를 자동 처리할 수도 있다.
# ---------------------------------------------------------
# Treatment(reference='No') → 기준 범주를 'No'로 지정
#
# model = smf.logit(
#     "Outcome ~ Age + C(Diabetes_History, Treatment(reference='No'))",
#     data=df
# ).fit()
#
# 그리고 Pre의 오즈비는 아래처럼 구할 수 있다.
#
# odds_ratio = np.exp(model.params["C(Diabetes_History, Treatment(reference='No'))[T.Pre]"])

# ② 예측확률 최대 환자의 Glucose (동률이면 ID 최소)
df["prob"] = model.predict(df)
top_row = df.sort_values(["prob", "ID"], ascending=[False, True]).iloc[0]

print("②", int(top_row["Glucose"]))

# ③ Threshold=0.5 기준 민감도(Sensitivity)
pred = (df["prob"] >= 0.5)

# confusion_matrix 구조
# [[TN, FP],
#  [FN, TP]]
# 민감도 = 실제 양성 중에서 모델이 양성으로 맞춘 비율 = TP / (TP + FN)

tn, fp, fn, tp = confusion_matrix(df["Outcome"], pred).ravel()
sensitivity = tp / (tp + fn)

print("③", round(sensitivity, 3))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_1samp
import statsmodels.formula.api as smf

# 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/BigData/Exam/11/3_2_promotion.csv')

# ① Page_Time의 평균이 300초와 같은지 t-검정 후 p-value
pvalue = ttest_1samp(df["Page_Time"], 300).pvalue
print("①", round(pvalue, 3))

# ② 수치형 변수들 간 상관계수 중 가장 큰 값 (자기 자신 제외)
corr = df.corr(numeric_only=True).abs()
np.fill_diagonal(corr.values, 0)
print("②", round(corr.max().max(),3))

# ③ Sales를 종속변수로 하는 다중선형회귀 후
#    p-value가 가장 작은 변수의 회귀계수 구하기
model = smf.ols(
    "Sales ~ Visit_Count + Page_Time + Click_Count + Ad_Spend + Conversion_Count",
    data=df
).fit()

pvalues = model.pvalues
target_var = pvalues.idxmin()
coef = model.params[target_var]

print("③", round(coef, 3))
```
