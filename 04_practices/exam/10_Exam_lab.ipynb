# 제1유형

1-1) 제공된 데이터 math.csv는 학생들의 수학 문제 풀이 결과를 기반으로 정답 여부를 포함한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Student ID : 학생 고유 번호
2. Topic : 전체 과목
3. Subtopic : 세부 과목(서브주제)
4. Answer : 문제 풀이 결과 (1: 정답, 0: 오답)

[수행 순서]
1. 각 서브 주제별로 정답률을 계산한다.
 ※ 정답률 = 정답 수 / 전체 응답 수
2. 정답률을 기준으로 내림차순으로 정렬한다.
3. 정답률이 세 번째로 높은 서브주제의 정답률을 구한다.

[문제]
1. 수행 순서 3)에서 계산한 값을 입력하시오.
☞ [제출 형식] 반올림하여 소수 셋째자리까지 작성

```python
import pandas as pd

df = pd.read_csv('../data/10/1_1_math.csv')

# 여기에 코드를 작성하시오.
# print(df.head(10))

processed = df.groupby('Subtopic')['Answer'].agg(['count', 'sum']) # groupby, agg 함께 사용하기
correct_rate = processed['sum']/processed['count'] # 분모 분자 헷갈리지 않게 주의
# print(correct_rate)
ans = round(correct_rate.sort_values(ascending=False).iloc[2],3)
print("ans: ", ans)

```

1-2) 제공된 데이터 sales.csv는 각 주문별 날짜, 카테고리, 매출 정보를 포함한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. order_date : 주문 날짜
2. category : 판매된 제품 카테고리
3. price : 해당 주문의 매출 금액

[수행 순서]
1. order_date 컬럼에서 연도(year)와 월(month)을 추출한다.
2. 연도와 월별로 price를 합산하여 총 매출액을 집계한 후, 두번째로 높은 총 매출액을 구한다.
3. 매출 기준 네 번째로 높은 연도/월을 찾고 해당 연도/월의 카테고리별 총 매출액을 집계한 뒤, 그 중 가장 높은 총 매출액을 구한다.

[문제]
1. 수행 순서 2)에서 계산한 총 매출액을 입력하시오.
 ☞ [제출 형식] 버림하여 정수(integer)로 작성
2. 수행 순서 3)에서 계산한 총 매출액을 입력하시오.
  ☞ [제출 형식] 버림하여 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv('../data/10/1_2_sales.csv')

# 여기에 코드를 작성하시오.

# print(df.head(10))
df['order_date'] = pd.to_datetime(df['order_date']) # to_datetime은 판다스 거 
df['year'] = df['order_date'].dt.year # year, month 가져오는 건 dt 거
df['month'] = df['order_date'].dt.month

total = df.groupby(['year','month'])['price'].sum()
# print(total)
second = total.sort_values(ascending=False).iloc[1]
# print(second)
ans1 = int(second)
print("ans1: ", ans1)

# print(total.sort_values(ascending=False)) # 2024 11

category_total = df[(df['year']==2024) & (df['month']==11)].groupby('category')['price'].sum()
# print(category_total)

ans2 = int(category_total.max())
print("ans2: ", ans2)
```

1-3) 제공된 데이터 spam.csv는 각 메시지와 해당 메시지가 스팸인지 여부를 포함한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Category : 메시지 분류 (ham : 일반 메시지, spam : 스팸 메시지)
2. Message : 문자 메시지 내용

[수행 순서]
1. Message 컬럼에서 공백(띄어쓰기, 줄바꿈 등)을 기준으로 단어 수를 세어 count라는 새로운 칼럼을 생성한다.
2. Category별(spam, ham)로 단어 수(count)의 평균을 계산한다.
3. 두 평균값의 차이를 절댓값으로 구한다.

[문제]
1. 수행 순서 3)에서 계산한 두 평균값의 차이의 절댓값을 입력하시오.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/10/1_3_spam.csv")

# 여기에 코드를 작성하시오.

# print(df.head())
def word_count(message):
    return len(message.split()) # df에 바로 split 불가능, 함수 만들기

# print(word_count("i love you k k k"))
df['count'] = df['Message'].apply(word_count)
# print(df)

avg_cnt = df.groupby('Category')['count'].mean()
# print(avg_cnt)

abs_diff = avg_cnt.iloc[1] - avg_cnt.iloc[0]
ans = round(abs_diff,3)

print("ans: ", ans)
```

# 제2유형

제공된 데이터(gas_train.csv)는 지역 상권에 대한 정보를 담고 있다. 이를 바탕으로 가스 총 사용량(GAS_TOTL)를 예측하는 회귀 모델을 개발하고, 가장 우수한 모델을 평가 데이터(gas_test.csv)에 적용하여 생존 예측 결과를 도출하시오. 예측 결과는 아래의 [제출 형식]을 준수하여, CSV 파일로 생성하는 코드를 제출하시오. (단, GAS_TOTL에는 결측치가 존재했으며, 모든 결측치는 0으로 대체된 상태이다.)
* 예측결과는 RMSE(Root Mean Squared Error) 평가지표에 따라 평가함

[제공 데이터]
1. gas_train.csv : 학습용 데이터 (800건)
2. gas_test.csv : 평가용 데이터 (200건, GAS_TOTL 컬럼 미제공)

[데이터 컬럼]
1. Commercial_District : 상업 지역 구분 (A, B, C, D, E)
2. Energy_Usage : 에너지 사용량
3. Population_Density : 인구 밀도
4. Vehicle_Count : 차량 수
5. Zoning_Score_Area : 구역 점수
6. Transport_Index : 교통 지수
7. GAS_TOTL : 가스 총 사용량

[제출 형식]
1. 제출 파일명 : result.csv
2. 제출 컬럼명 : pred
3. 예측 결과 개수 : 200




```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('../data/10/2_1_gas_train.csv')
test = pd.read_csv('../data/10/2_1_gas_test.csv')

# 답안 제출 참고
# 아래 코드는 예시이며 변수명 등 개인별로 변경하여 활용
# pd.DataFrame변수.to_csv('result.csv', index=False)

# 여기에 코드를 작성하시오.

# print(train.info())
# print(test.info())

# print(train.isnull().sum())
# print(test.isnull().sum())

except_zero = train['GAS_TOTL'][train['GAS_TOTL']!=0.000000]
train['GAS_TOTL'] = train['GAS_TOTL'].replace(0.000000, except_zero.mean())
# print(train['GAS_TOTL'].value_counts()) 
#0으로 채워진 결측값들 평균으로 대체

X = train.drop('GAS_TOTL', axis=1)
y = train['GAS_TOTL']

X_full = pd.concat([X, test], axis=0)
# print(X_full.shape)

X_full = pd.get_dummies(X_full)
# print(X_full)

X_train = X_full[:train.shape[0]]
test = X_full[train.shape[0]:]

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, random_state=0)
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)

# from sklearn.ensemble import RandomForestRegressor
# model = RandomForestRegressor()
# model.fit(X_train, y_train)

# from xgboost import XGBRegressor
# model = XGBRegressor()
# model.fit(X_train, y_train)

from lightgbm import LGBMRegressor
model = LGBMRegressor(n_estimators=100, max_depth=4)
model.fit(X_train, y_train)

from sklearn.metrics import root_mean_squared_error, r2_score
y_pred = model.predict(X_val)
rmse = root_mean_squared_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)
print(rmse, r2)

y_pred = model.predict(test)
result = pd.DataFrame(y_pred, columns=['pred'])
print(result)
result.to_csv('result_10.csv', index=False)
```

# 제3유형

3-1) 제공된 데이터는 특정 기업 직원의 퇴사 여부에 영향을 주는 정보를 포함한 자료이다. 아래 데이터를 이용하여 퇴사여부를 예측하고자 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오.

[제공 데이터]
1. Age : 직원 나이
2. MonthlyIncome : 월급
3. OverTime : 야근 여부 (0: 없음, 1: 있음)
4. Attrition : 퇴사 여부 (0: 재직, 1: 퇴사)

[문제]
1. Age, MonthlyIncome, OverTime을 독립변수로 사용하여 로지스틱 회귀분석을 수행하고, 가장 통계적으로 유의한 변수에 대하여 회귀계수를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①에서 구축한 모형을 이용하여, 나이가 1살 증가할 때마다 퇴사할 오즈비(Odds Ratio)를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ①에서 구축한 모형을 이용하여, 아래 새로운 직원 정보가 주어졌을 때, 퇴사할 확률을 구하여라.
  (Age=45, MonthlyIncome=4500, OverTime=1)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/10/3_1_employee.csv")

# 여기에 코드를 작성하시오.
import statsmodels.formula.api as smf

model = smf.logit('Attrition ~ Age + MonthlyIncome + OverTime', data=df).fit()
# print(model.summary())

ans1 = round(model.params['MonthlyIncome'],3)
print("ans1: ", ans1)

import numpy as np
odds_ratio = float(np.exp(model.params['Age']))
ans2 = round(odds_ratio,3)
print("ans2: ", ans2)

new_data = pd.DataFrame({'Age': [45], 'MonthlyIncome': [4500], 'OverTime': [1]})
pred = model.predict(new_data)
ans3 = round(pred[0],3)
print("ans3: ", ans3)
```

3-2) 제공된 데이터 house.csv는 주택의 다양한 특성과 판매 가격 정보를 포함한 자료이다. 아래 데이터를 이용하여 주택 가격을 예측하고자 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 모델은 절편을 포함한다.)

[제공 데이터]
1. Area : 주택 면적
2. Stories : 층 수
3. Wall : 벽 마감재 사용 여부 (1: 예, 0: 아니오)
4. Price : 주택 가격

[문제]
1. Area, Stories, Wall을 독립변수로 사용하여 다중선형회귀분석을 수행하고, 변수들의 회귀계수 합(절편 제외)을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①번에서 통계적으로 유의하지 않은 변수는 제거하고, 남은 유의한 변수들만 포함하여 다중회귀모형을 다시 적합한 후 결정계수(𝑹^𝟐) 값을 구하시오..
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②번에서 만든 최종 모델을 이용하여, 아래 새로운 주택의 정보를 입력했을 때의 예상 가격을 구하여라.
  (Area = 120, Stories = 2.5, Wall = 1)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/10/3_2_house.csv")

# 여기에 코드를 작성하시오.

import statsmodels.formula.api as smf

model = smf.ols('Price ~ Area + Stories + Wall', data=df).fit()
# print(model.summary())

ans1 = round(model.params['Area'] + model.params['Stories'] + model.params['Wall'],3)
print("ans1: ", ans1)

model2 = smf.ols('Price ~ Area + Stories', data=df).fit()
# print(model2.summary())
ans2 = round(model.rsquared,3)
print("ans2: ", ans2)

new_data = pd.DataFrame({'Area' : [120], 'Stories' : [2.5], 'Wall' : [1]})
pred = model2.predict(new_data)
ans3 = round(pred[0],3)
print("ans3: ", ans3)
```
