1-2) 제공된 데이터 tourism.csv는 실제 국가별 관광 목적 및 공무 목적 입국자 수에 대한 정보를 포함한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. country : 국가명
2. tourism : 관광 목적 입국자 수
3. official : 공무 목적 입국자 수

[수행 순서]
1. '총 입국 인원' 과 '관광객 비율' 파생변수를 생성한다.
 ※ 총 입국 인원 = 관광 입국 인원 + 공무 입국 인원
 ※ 관광객 비율 = 관광 입국 인원 / 총 입국 인원
2. 관광객 비율이 두 번째로 높은 나라의 관광 입국 인원을 A라고 한다.
3. 총 입국 인원이 두 번째로 많은 나라의 공무 입국 인원을 B라고 한다.

[문제]
1. 수행 순서 2)와 3)에서 구한 A, B의 값의 합(A+B)을 입력하시오.
 ☞ [제출 형식] 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv('../data/8/1_2_tourism.csv')

# 여기에 코드를 작성하시오.

# print(df.head(10))

df['총 입국 인원'] = df['tourism'] + df['official']
df['관광객 비율'] = df['tourism']/df['총 입국 인원']
# print(df)

a_value = df['관광객 비율'].sort_values(ascending=False).iloc[1]
A = df[df['관광객 비율']==a_value]['tourism'].iloc[0]
# print(A)

b_value = df['총 입국 인원'].sort_values(ascending=False).iloc[1]
B = df[df['총 입국 인원'] == b_value]['official'].iloc[0]
# print(B)

ans = A+B
print("ans: ", ans)
```

아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Co : 원재료에 포함된 탄소 농도
2. Nmch : 설비 성능 지표

[수행 순서]
1. Co 컬럼과 Nmch 컬럼에 대해 최대-최소 정규화(Min-Max Scaling) 를 수행한다.
2. 정규화된 Co 컬럼의 표본표준편차를 A, 정규화된 Nmch 컬럼의 표본표준편차를 B라고한다.

[문제]
1. 수행 순서 2)에서 구한 A, B의 값의 차이(A-B)를 입력하시오.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv('../data/8/1_3_analysis.csv')

# 여기에 코드를 작성하시오.

# print(df.head(10))

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# 정규화도 칼럼별로 진행할 것
processed_co = scaler.fit_transform(df[['Co']])
processed_nmch = scaler.fit_transform(df[['Nmch']]) # 에러나면 당황하지 말고 대괄호 2번
# print(processed)
# 판다스: 표본표준편차, 넘파이: 모표준편차
A = processed_co.std()
B = processed_nmch.std()
# print(A, B)

ans = round(A-B, 3)
print("ans: ", ans)
```

# 제2유형

제공된 데이터(subway_train.csv)는 특정 도시 지하철역의 환경 및 시설 정보 담고 있다. 이를 바탕으로 하루 평균 방문 인원수(Passengers)를 예측하는 회귀 모델을 개발하고, 가장 우수한 모델을 평가 데이터(subway_test.csv)에 적용하여 예측 결과를 도출하시오. 예측 결과는 아래의 [제출 형식]을 준수하여, CSV 파일로 생성하는 코드를 제출하시오.
 * 예측결과는 MAE(Mean Absolute Error) 평가지표에 따라 평가함

[제공 데이터]
1. subway_train.csv : 학습용 데이터 (824건)
2. subway_test.csv : 평가용 데이터 (312건, Passengers 미제공)

[데이터 컬럼]
1. StationID: 지하철역 고유 ID
2. Line: 노선 이름 (Line1 ~ Line5)
3. Temperature: 평균 기온
4. Rainfall_mm: 강수량(mm)
5. Is_Holiday: 공휴일 여부 (0: 평일, 1: 공휴일)
6. Nearby_Shops: 상업시설 개수
7. Transfer_Stations: 환승역 수 (0~3)
8. Bus_Connections: 연결된 버스 노선 수
9. Passengers: 하루 방문 승객 수

[제출 형식]
1. 제출 파일명 : result.csv
2. 제출 컬럼명 : pred
3. 예측 결과 개수 : 312

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('../data/8/2_1_subway_train.csv')
test = pd.read_csv('../data/8/2_1_subway_test.csv')

# 답안 제출 참고
# 아래 코드는 예시이며 변수명 등 개인별로 변경하여 활용
# pd.DataFrame변수.to_csv('result.csv', index=False)

# 여기에 코드를 작성하시오.

# print(train.info())
# print(test.info())

# print(train.isnull().sum())
# print(test.isnull().sum())

X = train.drop(['Passengers', 'StationID'], axis=1)
y = train['Passengers']
test = test.drop('StationID', axis=1)

X_full = pd.concat([X, test], axis=0)
# print(X_full.shape)

X_full = pd.get_dummies(X_full)
# print(X_full)

X_train = X_full[:train.shape[0]]
test = X_full[train.shape[0]:]

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, random_state=0)
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)

# from sklearn.ensemble import RandomForestRegressor
# model = RandomForestRegressor()
# model.fit(X_train, y_train)

# from xgboost import XGBRegressor
# model = XGBRegressor()
# model.fit(X_train, y_train)

from lightgbm import LGBMRegressor
model = LGBMRegressor(n_estimators=100, max_depth=4)
model.fit(X_train, y_train)

from sklearn.metrics import mean_absolute_error, r2_score
y_pred = model.predict(X_val)
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)
print(mae, r2)

y_pred = model.predict(test)
result = pd.DataFrame(y_pred, columns= ['pred'])
print(result)
result.to_csv('result_8.csv', index=False)
```

3-1) 제공된 데이터(churn.csv)는 통신 서비스 고객의 사용 특성과 이탈 여부를 포함하고 있으며, 아래 데이터를 이용하여 고객 이탈 여부를 예측하고자 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오.

[제공 데이터]
1. AvgCallMinutes : 고객의 월 평균 통화 시간 (분)
2. ContractMonths :	계약 유지 개월 수
3. InternationalPlan : 국제전화 요금제 가입 여부
4. TechSupport : 기술지원 서비스 가입 여부
5. Calls : 월 평균 전화 건수
6. Churn : 고객 이탈 여부 (1: 이탈, 0: 유지)

[문제]
1. 위 데이터로 로지스틱 회귀분석을 수행하였을 때, 유의수준 0.05 기준으로 유의하지 않은 독립변수의 개수를 구하여라.
 ☞ [제출 형식] 정수(integer)로 작성
2. ①에서 유의하지 않은 독립변수들을 제거한 후 회귀모델을 재학습하고, 회귀계수들의 평균값(절편 제외)을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②에서 구축한 모델을 활용하여, Calls 값이 5 증가할 때 오즈비가 몇 배 증가하는지를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/8/3_1_churn.csv")

# 여기에 코드를 작성하시오.
import statsmodels.formula.api as smf
model = smf.logit('Churn ~ AvgCallMinutes + ContractMonths + InternationalPlan + TechSupport + Calls', data=df).fit()
# print(model.summary())
ans1 = 2
print("ans1: ",ans1)

model2 = smf.logit('Churn ~ InternationalPlan + TechSupport + Calls', data=df).fit()
# print(model2.summary())
ans2 = round((-0.4970 - 0.6359 + 0.2050)/3, 3)
print("ans2: ", ans2)

import numpy as np
odds_ratio = float(np.exp(model2.params['Calls']*5)) # 5가 증가할 때... 는 exp괄호 안에 *5(맨 뒤에 곱하기 5배 아님)
ans3 = round(odds_ratio,3)
print("ans3: ", ans3)
```

3-2) 제공된 데이터(piq.csv)는 개인의 지능지수(PIQ)와 신체 및 두뇌 관련 정보를 포함하고 있으며, 아래 데이터를 이용하여 지능지수를 예측하고자 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 모델은 절편을 포함한다.)

[제공 데이터]
1. PIQ: 지능지수
2. BRAIN: 뇌 용적 (cm³)
3. HEIGHT: 신장 (inch)
4. WEIGHT: 체중 (pound)

[문제]
1. BRAIN, HEIGHT, WEIGHT를 독립변수로 사용하여 다중선형회귀분석을 수행하고, 이때 p-value가 가장 작은 독립변수의 회귀계수 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①에서 구축한 모델의 결정계수(R-Sqaured) 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ①에서 구축한 모델을 이용하여, 아래 새로운 데이터 정보를 입력했을 때 예상 PIQ를 구하여라.
(BRAIN=100, HEIGHT=72, WEIGHT=128)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/8/3_2_piq.csv")

# 여기에 코드를 작성하시오.

import statsmodels.formula.api as smf
model = smf.ols('PIQ ~ BRAIN + HEIGHT + WEIGHT', data=df).fit()
# print(model.summary())

ans1 = round(model.params['WEIGHT'],3)
print("ans1: ", ans1)

ans2 = round(model.rsquared,3)
print("ans2: ",ans2)

new_data = pd.DataFrame({'BRAIN':[100], 'HEIGHT':[72], 'WEIGHT':[128]})
pred = model.predict(new_data)
ans3 = round(pred.iloc[0],3)
print("ans3: ", ans3)
```
