```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('/content/drive/MyDrive/BigData/Exam/7/2_1_usage_train.csv')
test = pd.read_csv('/content/drive/MyDrive/BigData/Exam/7/2_1_usage_test.csv')

# X, Y 데이터 셋 분리
X = train.drop(['이용금액'], axis=1)
y = train['이용금액']

X_full = pd.concat([X, test], axis=0)
#print(X_full.shape)

# 범주형 변수 인코딩 - One-Hot Encoding
X_full = pd.get_dummies(X_full)
#print(X_full)

# 학습, 검증 데이터 셋 분할
X_train = X_full[:train.shape[0]]
X_test = X_full[train.shape[0]:]
#print(X_train.shape, X_test.shape)

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2)
#print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)

# Randomforest 활용 모델 학습
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)

# RMSE, R2 Score 활용 평가
from sklearn.metrics import root_mean_squared_error, r2_score
y_pred = model.predict(X_val)
rmse = root_mean_squared_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)
print(rmse, r2)

# test 데이터 예측 및 결과 저장 (pred 1개 컬럼)
y_pred = model.predict(X_test)
result = pd.DataFrame(y_pred, columns=['pred'])
result.to_csv('result.csv', index=False)
```

3-1) 제공된 데이터(student_fitness.csv)는 개인의 신체 측정값과 성별 여부를 포함한 자료이다. 성별(gender) 을 예측하고자 한다. 전체 300명 중 앞의 210명은 학습용(train) 데이터로 사용하고, 뒤의 90명은 시험용(test) 데이터로 사용한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 벌점화(penalty)는 부여하지 않는다.)

[제공 데이터]
1. height : 개인의 키 정보 (cm)
2. weight :	개인의 몸무게 정보 (kg)
3. grip : 악력 측정값 (kgf)
4. shoulder : 어깨둘레 (cm)
5. age : 개인의 실제 연령 정보 (세)
6. gender : 성별 (1 : 남성, 0 : 여성)

[문제]
1. weight 단일 변수를 독립변수로 하여 학습용(train) 데이터를 이용해 로지스틱 회귀분석을 수행하고, weight가 1 증가할 때 성별이 남자일 오즈비가 얼마나 증가하는지 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. height, weight, grip, shoulder, age를 독립변수로 하여 로지스틱 회귀분석을 수행하고, 구축한 모형의 잔차이탈도(Deviance)를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②에서 구축한 모형을 시험용(test) 데이터에 적용하여 예측값을 구하고, 오류율(Error Rate)을 구하여라.
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/7/3_1_student_fitness.csv")

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
from sklearn.metrics import accuracy_score

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/7/3_1_student_fitness.csv")

# 학습/평가 분할
train, test = df.iloc[:210], df.iloc[210:]

# ① 몸무게(weight) 1 증가 시 성별(gender)에 대한 오즈비 계산
model = smf.logit("gender ~ weight", data=train).fit()
odds_ratio = float(np.exp(model.params["weight"]))
print("① 오즈비:", round(odds_ratio, 3))

# ② 잔차이탈도 = -2 * 로그우도(Log-Likelihood)
# height, weight, grip, shoulder, age 를 모두 사용한 로지스틱 회귀모형 적합
model_full = smf.logit("gender ~ height + weight + grip + shoulder + age", data=train).fit()
deviance = -2 * float(model_full.llf)
print("② 잔차이탈도:", round(deviance, 3))

# ③ test 데이터에서 오류율(Error Rate) 계산
proba = model_full.predict(test[['height','weight','grip','shoulder','age']])
# proba = model_full.predict(test)
# ↑ 겉보기엔 동일하게 동작하지만,
#    predict()는 모델에 사용된 독립변수 컬럼만 필요로 하므로
#    test에 불필요한 컬럼이 포함되거나 순서가 달라지면 예측 결과가 달라질 수 있음.
#    → 따라서 안전하게 test[['height','weight','grip','shoulder','age']] 로 지정하는 것이 정석.
pred = (proba >= 0.5).astype(int)
error_rate = 1 - accuracy_score(test['gender'], pred)
print("③ 오류율:", round(error_rate, 3))
```

[문제]

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/7/3_2_car_efficiency.csv")

# 여기에 코드를 작성하시오.
```

```python
import pandas as pd
import statsmodels.formula.api as smf

df = pd.read_csv("/content/drive/MyDrive/BigData/Exam/7/3_2_car_efficiency.csv")

# ① 연비(FuelEfficiency)와 각 독립변수 간의 상관계수 확인
corrs = df.corr(numeric_only=True)["FuelEfficiency"].drop("FuelEfficiency")
max_corr_var = corrs.abs().idxmax()
ans1 = round(corrs[max_corr_var], 3)

# Notice) 상관계수는 '절댓값'이 클수록 연비와의 관계가 강하다는 의미이다.
# 예) -0.8 은 +0.5 보다 연비와 더 강한 관련성이 있다.
print("① 상관계수:", ans1)

# ② 다중선형회귀모형 적합 및 결정계수(R-Square) 계산
model = smf.ols("FuelEfficiency ~ EngineSize + VehicleWeight + AeroCoefficient + TireResistance", data=df).fit()
ans2 = round(model.rsquared, 3)
print("② 결정계수:", ans2)

# ③ p-value가 가장 큰(=유의성이 가장 낮은) 독립변수의 p-value 값
pvals = model.pvalues
ans3 = round(pvals.max(), 3)
print("③ p-value:", ans3)
```
