1-1) 제공된 데이터 emission.csv는 2001년 ~ 2021년까지 전세계 국가별 온실가스 배출량을 포함한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. year : 연도
2. country : 국가명
3. emission : 온실가스 배출량

[수행 순서]
1. 연도별 온실가스 배출량 1위를 가장 많이 한 국가를 구한다.
2. 1에서 구한 연도별 1위 국가들 중, 가장 많이 1위 한 국가를 구하고 해당 국가가 1위를 한 연도 횟수를 구한다.
3. 전체 기간(2001~2021)의 배출량 연평균 성장률(CAGR)을 계산한다.
※ 𝑪𝑨𝑮𝑹=((𝟐𝟎𝟐𝟏년도 총 배출량)/(𝟐𝟎𝟎𝟏년도 총 배출량))^((𝟏/(𝟐𝟎𝟐𝟏−𝟐𝟎𝟎𝟏)))−𝟏

[문제]
1. 수행 순서 2)에서 계산한 1위 횟수를 입력하시오.
 ☞ [제출 형식] 정수(integer)로 작성
2. 수행 순서 3)에서 계산한 CAGR을 입력하시오.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv('../data/11/1_1_emission.csv')

# 여기에 코드를 작성하시오.

# print(df.head(10))
df_index = df.set_index('country')
highest_countries = df_index.groupby(['year'])['emission'].idxmax()
# print(highest_countries)
highest = highest_countries.value_counts()
# print(highest)

ans1 = 12
print("ans1: ", ans1)

total_2021 = df[df['year'] == 2021].groupby('year')['emission'].sum().iloc[0]
total_2001 = df[df['year'] == 2001].groupby('year')['emission'].sum().iloc[0]
# print(total_2001)
CAGR = (total_2021/total_2001)**(1/(2021-2001))-1
ans2 = round(CAGR,3)
print("ans2: ", ans2)
```

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcmd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이버 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv('../data/11/2_1_game_level_train.csv')
test = pd.read_csv('../data/11/2_1_game_level_test.csv')

# 답안 제출 참고
# 아래 코드는 예시이며 변수명 등 개인별로 변경하여 활용
# pd.DataFrame변수.to_csv('result.csv', index=False)

# 여기에 코드를 작성하시오.

# print(train.info())
# print(test.info())

# print(train.isnull().sum())
# print(test.isnull().sum())

X = train.drop('Game_Level', axis=1)
y = train['Game_Level']

X_full = pd.concat([X,test], axis=0)
# print(X_full.shape)

X_full = pd.get_dummies(X_full)
# print(X_full)

X_train = X_full[:train.shape[0]]
test = X_full[train.shape[0]:]

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, stratify=y, random_state=0)
# print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)

# from sklearn.ensemble import RandomForestClassifier
# model = RandomForestClassifier()
# model.fit(X_train, y_train)

# from xgboost import XGBClassifier
# model = XGBClassifier()
# model.fit(X_train, y_train)

from lightgbm import LGBMClassifier
model = LGBMClassifier(n_estimators=300, max_depth=8)
model.fit(X_train, y_train)

from sklearn.metrics import f1_score, accuracy_score
y_pred = model.predict(X_val)
f1 = f1_score(y_val, y_pred, average='macro')
acc = accuracy_score(y_val, y_pred)
print(f1, acc)

y_pred = model.predict(test)
result = pd.DataFrame(y_pred, columns=['pred'])
print(result)
result.to_csv('result_11.csv', index=False)
```

3-1) 제공된 데이터(diabetes.csv)는 환자의 건강 정보와 당뇨병 발생 여부를 포함한 자료이다. 아래 데이터를 이용하여 로지스틱 회귀 분석 및 성능 평가를 수행하고자 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 벌점화(penalty)는 부여하지 않는다.)

[제공 데이터]
1. ID : 환자 식별 번호
2. Age : 연령
3. Diabetes_History : 당뇨 가족력 (No, Pre, Yes)
4. Glucose : 공복 혈당
5. Outcome : 당뇨병 발생 여부 (0: 정상, 1: 당뇨)

[문제]
1. 독립변수는 Age, Diabetes_History를 사용하여 로지스틱 회귀분석을 수행하시오. 이때 Diabetes_History의 Pre(당뇨 전단계)에 대한 오즈비(Odds Ratio) 를 구하여라. (단, 기준 그룹은 No로 한다.)
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①에서 구축한 모델로 전체 데이터의 Outcome=1 발생 확률을 예측하시오. 그 중 발생 확률이 가장 높은 환자의 Glucose 값을 구하여라. (확률이 동일한 환자가 여러 명이면 ID가 가장 작은 환자를 선택)
 ☞ [제출 형식] 정수(integer)로 작성
3. ①에서 구축한 모델에 대해 임계값(Threshold)=0.5로 분류했을 때의 민감도(Sensitivity) 를 구하여라. (Outcome=1을 양성으로 간주)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/11/3_1_diabetes.csv")

# 여기에 코드를 작성하시오.
import statsmodels.formula.api as smf
model = smf.logit('Outcome ~ Age + Diabetes_History', data=df).fit()
# print(model.summary())
import numpy as np
odds_ratio = float(np.exp(model.params["Diabetes_History[T.Pre]"]))
ans1 = round(odds_ratio,3)
print("ans1: ", ans1)

proba = model.predict(df)
# print(pred.idxmax())
ans2 = round(df['Glucose'].iloc[proba.idxmax()],3)
print("ans2: ", ans2)

pred2 =(proba>=0.5)
# 혼동행렬 다시 암기하기
# [[TN, FP]
#  [FN, TP]]
# 민감도: 실제 양성 중에 모델이 양성으로 맞춘 거... TP/(TP+FN)
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(df["Outcome"], pred2).ravel()
sensitivity = tp/(tp+fn)
ans3= round(sensitivity,3)
print("ans3: ", ans3)
```

```python
import pandas as pd

df = pd.read_csv('../data/11/3_2_promotion.csv')

# 여기에 코드를 작성하시오.

from scipy import stats
t, p = stats.ttest_1samp(df['Page_Time'], 300) # 뒤에 .mean() 없이
ans1 = round(p,3)
print("ans1: ", ans1)

# print(df.corr().abs())
ans2 = round(0.369826,3)
print("ans2: ", ans2)

import statsmodels.formula.api as smf
model = smf.ols('Sales ~ Visit_Count + Page_Time + Click_Count + Ad_Spend + Conversion_Count', data=df).fit()
# print(model.summary())
target_var =  model.pvalues.idxmin() #p값이 0인 게 두개...소수점 3째자리까지만 보여줌, t가 더 큰 걸 보자
ans3 = round(model.params[target_var],3)
print("ans3: ", ans3)
```
