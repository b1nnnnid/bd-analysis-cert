1-3) 제공된 데이터 smartfarm_env.csv는 스마트 온실 환경 모니터링 시스템에서 수집된 생육 환경 데이터를 포함한다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 [제출 형식]에 맞춰 입력하시오.

[수행 순서]
1. soil_moisture 컬럼에서 IQR 기준으로 이상값을 판단한다.
 ※ 이상값 기준 : 하한선(Q1-1.5 x IQR), 하한선(Q3+1.5 x IQR)
2. 위 기준을 벗어나는 이상값들의 개수를 구하여 입력한다.

[제출 형식]
정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv('../data/7/1_3_smartfarm_env.csv')

# 여기에 코드를 작성하시오.

Q1 = df['soil_moisture'].quantile(0.25) # 한 IQR이 아닌 Q1, Q3을 개별적으로 구하기(quantile)
Q3 = df['soil_moisture'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[(df['soil_moisture'] < Q1-1.5*IQR) | (df['soil_moisture'] > Q3+1.5*IQR)]
ans = len(outliers) # 단순 행 수 구할 땐 len 도 좋다!

print("ans: ", ans)
```

7. 평균체류시간 : 1회 방문 시 평균 체류 시간
8. 공휴일여부, 이벤트진행 : 해당 1, 미해당 0

# 제3유형

3-1) 제공된 데이터(student_fitness.csv)는 개인의 신체 측정값과 성별 여부를 포함한 자료이다. 성별(gender) 을 예측하고자 한다. 전체 300명 중 앞의 210명은 학습용(train) 데이터로 사용하고, 뒤의 90명은 시험용(test) 데이터로 사용한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 벌점화(penalty)는 부여하지 않는다.)

[제공 데이터]
1. height : 개인의 키 정보 (cm)
2. weight :	개인의 몸무게 정보 (kg)
3. grip : 악력 측정값 (kgf)
4. shoulder : 어깨둘레 (cm)
5. age : 개인의 실제 연령 정보 (세)
6. gender : 성별 (1 : 남성, 0 : 여성)

[문제]
1. weight 단일 변수를 독립변수로 하여 학습용(train) 데이터를 이용해 로지스틱 회귀분석을 수행하고, weight가 1 증가할 때 성별이 남자일 오즈비가 얼마나 증가하는지 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. height, weight, grip, shoulder, age를 독립변수로 하여 로지스틱 회귀분석을 수행하고, 구축한 모형의 잔차이탈도(Deviance)를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②에서 구축한 모형을 시험용(test) 데이터에 적용하여 예측값을 구하고, 오류율(Error Rate)을 구하여라.
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/7/3_1_student_fitness.csv")

# 여기에 코드를 작성하시오.

train = df.iloc[:210]
test = df.iloc[210:]
# print(train.shape, test.shape)

import statsmodels.formula.api as smf
import numpy as np
model = smf.logit('gender ~ weight', data=train).fit() # train 데이터 활용
# print(model.summary())
odds_ratio = float(np.exp(model.params['weight'])) # 오즈비 계산식 암기하기
ans1 = round(odds_ratio,3)
print("ans1: ", ans1)

model2 = smf.logit('gender ~ height + weight + grip + shoulder + age', data=train).fit()
ans2 = round(-2*model2.llf, 3) # 로그우도 구할 땐 그냥 모델 뒤에 .llf, 잔차이탈도는 곱하기 -2
print("ans2: ", ans2)

from sklearn.metrics import accuracy_score
proba = model2.predict(test[['height','weight','grip','shoulder','age']])
pred = (proba > 0.5).astype(int) # 확률값들 이진으로 변환하기

acc = accuracy_score(test['gender'], pred) # 반드시 타겟 칼럼 지정 
ans3 = round(1-acc,3) #에러율은 그냥 1-acc
print("ans3: ", ans3)
```

3-2) 제공된 데이터(car_efficiency.csv)는 자동차의 연비(FuelEfficiency) 와 엔진 및 차량 특성 정보를 포함하고 있으며, 아래 데이터를 이용하여 연비를 예측한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 모델은 절편을 포함한다.)

[제공 데이터]
1. FuelEfficiency : 연비 (km/L)
2. EngineSize : 엔진 배기량 (L)
3. VehicleWeight : 차량 중량 (kg)
4. AeroCoefficient : 공기저항 계수
5. TireResistance : 타이어 마찰계수

[문제]
1. FuelEfficiency와 각 독립변수의 상관계수를 계산하고, 가장 높은 상관관계를 갖는 변수의 상관계수 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. EngineSize, VehicleWeight, AeroCoefficient, TireResistance 를 독립변수로 하여 다중선형회귀모형을 적합하고, 모형의 결정계수(R-Square)를 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②에서 구축한 모형에서 p-value가 가장 높은 변수의 값을 구하여라.
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.read_csv("../data/7/3_2_car_efficiency.csv")

# 여기에 코드를 작성하시오.

corr = df.corr()
corr['FuelEfficiency'] = corr['FuelEfficiency'].drop('FuelEfficiency', axis=0)
# print(corr['FuelEfficiency'])
ans1 = round(-0.480033,3)
print("ans1: ", ans1)

import statsmodels.formula.api as smf
model = smf.ols('FuelEfficiency ~ EngineSize + VehicleWeight + AeroCoefficient + TireResistance', data=df).fit()
# print(model.summary())
ans2 = round(model.rsquared,3)
print("ans2: ", ans2)

ans3 = round(max(model.pvalues),3)
print("ans3: ", ans3)
```
