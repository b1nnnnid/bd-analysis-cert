제공된 데이터는 특정 수업을 들은 학생들의 성적과 성적에 영향을 주는 정보들을 포함한 자료이다. 아래 데이터를 이용하여 성적을 예측하고자 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 모델은 절편을 포함한다.)

[제공 데이터]
- 공부시간, 과제점수, 모의고사, 출석률, 수면시간, 성적

[문제]
1. 공부시간, 과제점수, 모의고사, 출석률, 수면시간을 독립변수로 사용하여 다중선형회귀분석을 수행하고, 이때 p-value가 가장 작은 독립변수의 회귀계수 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. ①의 결과를 바탕으로 유의수준 0.05 기준에서 유의하지 않은 독립변수를 제거한 후, 남은 변수만을 이용하여 다시 다중선형회귀분석을 수행하고 결정계수(R-squared) 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②에서 구축한 최종 모형을 이용하여, 아래 새로운 학생의 정보를 입력했을 때 예상 성적을 구하여라.
(공부시간=9, 과제점수=84, 모의고사=82, 출석률=95, 수면시간=7.0)
  ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.DataFrame({
    "공부시간": [6, 8, 10, 4, 7, 12, 9, 5, 11, 3, 14, 8, 6, 13, 10, 7, 9, 5, 12, 4, 11, 6, 8, 10],
    "과제점수": [78, 82, 90, 70, 76, 92, 85, 73, 94, 68, 96, 80, 77, 95, 88, 79, 84, 72, 93, 71, 91, 75, 83, 89],
    "모의고사": [74, 79, 86, 65, 72, 91, 83, 69, 88, 62, 93, 78, 73, 92, 85, 76, 82, 67, 90, 64, 89, 71, 80, 87],
    "출석률":   [92, 95, 96, 88, 93, 97, 95, 90, 96, 87, 98, 94, 92, 97, 96, 93, 95, 89, 97, 88, 96, 92, 94, 95],
    "수면시간": [7.5, 7.0, 6.8, 8.0, 7.2, 6.5, 7.1, 7.8, 6.7, 8.2, 6.4, 7.3, 7.6, 6.6, 6.9, 7.4, 7.0, 7.9, 6.5, 8.1, 6.6, 7.5, 7.2, 6.8],
    "성적": [76, 80, 88, 70, 75, 94, 86, 72, 92, 66, 97, 81, 77, 96, 89, 79, 85, 71, 95, 68, 93, 74, 82, 90]
})

# 여기에 코드를 작성하시오.

import statsmodels.formula.api as smf
model = smf.ols('성적 ~ 공부시간 + 과제점수 + 모의고사 + 출석률 + 수면시간', data=df).fit()
#print(model.summary())
ans1 = round(0.9168, 3) # 모의고사
print("ans1: ", ans1)

model2 = smf.ols('성적 ~ 공부시간 + 모의고사 + 출석률', data=df).fit() # 과제점수, 수면시간 빠짐
# print(model2.summary())
ans2 = 0.995
print("ans2: ", ans2)

new_data = pd.DataFrame({"공부시간": [9], "과제점수": [84], "모의고사": [82], "출석률": [95], "수면시간": [7.0]}) # 값은 무조건 리스트 형태로
pred = model2.predict(new_data)
ans3 = round(pred[0],3)
print("ans3: ", ans3)
```

# 풀이(#2)

```python
import statsmodels.formula.api as smf

m_full = smf.ols('성적 ~ 공부시간 + 과제점수 + 모의고사 + 출석률 + 수면시간', data=df).fit()
pvals = m_full.pvalues.drop('Intercept')
best_var = pvals.idxmin()
best_coef = m_full.params[best_var]
print("[1] p-value 최솟값 변수:", best_var)
print("[1] 그 변수의 회귀계수:", round(best_coef, 3))

keep_vars = pvals[pvals < 0.05].index.tolist()
formula_final = '성적 ~ ' + ' + '.join(keep_vars)
m_final = smf.ols(formula_final, data=df).fit()
print("[2] 최종 모형:", formula_final)
print("[2] R-squared:", round(m_final.rsquared, 3))

new_student = pd.DataFrame({
    "공부시간":[9], "과제점수":[84], "모의고사":[82],
    "출석률":[95], "수면시간":[7.0]
})
pred = m_final.predict(new_student)
print("[3] 예측 성적:", round(float(pred[0]), 3))
```
