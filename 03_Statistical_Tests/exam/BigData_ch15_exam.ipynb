문제1) 제공된 데이터는 특정 학교 학생들의 시험 점수를 나타내는 자료이며, 학생들의 시험 점수가 특정 기준 값과 일치하는지 검정하려고 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오.

[제공 데이터]
1. score : 시험 점수

[문제]
1. 학생들의 시험 점수가 정규성을 만족하는지 알아보기 위해 샤피로-윌크 테스트를 수행할 때 p-value 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. 학생들의 시험 점수가 80점과 같은지 검정하고자 한다. ①에서 구한 결과를 근거로 정규성을 만족하면 단일표본 t-검정, 정규성을 만족하지 않으면 윌콕슨 부호순위 검정을 수행하여 p-value 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②에서 사용한 검정방식을 활용하여, 학생들의 시험 점수가 80점보다 낮은지 검정을 수행하고 p-value 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.DataFrame({
    "score": [81, 83, 82, 80, 79, 82, 81, 84, 78, 82,
              80, 83, 81, 79, 82, 84, 80, 81, 83, 79,
              82, 81, 80, 84, 78, 82, 81, 83, 78, 82]
})

# 여기에 코드를 작성하시오.

from scipy import stats
stats.shapiro(df['score'])
ans1 = round(0.11485444996828698,3)
print("ans1: ", ans1) # 정규성 만족 -> 단일표본 t검정

stats.ttest_1samp(df['score'], 80, alternative='two-sided') # 양측 검정
ans2 = round(0.0012201432943160506,3) # 기각: 80과 같지 않다
print("ans2: ", ans2)

stats.ttest_1samp(df['score'], 80, alternative='less')
ans3 = round(0.9993899283528419,3) # 채택: 80보다 낮다
print("ans3: ", ans3)
```

# 풀이

```python
from scipy import stats

# ① Shapiro-Wilk 정규성 검정
w_stat, p_value_shapiro = stats.shapiro(df['score'])
print("1번 Shapiro-Wilk p-value:", round(p_value_shapiro, 3))

# ② 단일표본 t-검정 (양측)
t_stat_2, p_value_2 = stats.ttest_1samp(df['score'], 80)
print("2번 p-value:", round(p_value_2, 3))

# ③ 단일표본 t-검정 (단측: 평균 < 80)
t_stat_3, p_value_3 = stats.ttest_1samp(df['score'], 80, alternative='less')
print("3번 p-value:", round(p_value_3, 3))
```
