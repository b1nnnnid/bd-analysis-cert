문제1) 제공된 데이터는 2가지 약에 대한 자료이며, 두 신약의 효능 지속시간에 대해 검증하고자 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오. (단, 제공된 데이터는 정규성과 등분산성을 만족한다고 가정한다.)

[제공 데이터]
1. A : A약 복용 후, 효능 지속시간(분)
2. B : B약 복용 후. 효능 지속시간(분)

[문제]
1. 2가지 신약을 A집단과 B집단 두 개의 집단에 대하여 비교한 데이터라고 할 때, 효능 지속시간에 효과가 있는지 t-검정을 수행하고 p-value 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. 하나의 집단에 대하여 A약이 과거자료,  B약이 신약의 효능을 비교한 데이터라고 할 때, 효능 지속시간에 효과가 있는지 t-검정을 수행하고 p-value 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. ②의 조건에서 정규성을 만족하지 않는다고 가정하고, 윌콕슨 검정을 수행하고 p-value 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.DataFrame({
    "A": [81.5, 79.6, 81.9, 84.6, 79.3, 79.3, 84.7, 82.3, 78.6, 81.6,
          78.6, 78.6, 80.7, 74.3, 74.8, 78.3, 77.0, 80.9, 77.3, 75.8,
          84.4, 79.3, 80.2, 75.7, 78.4, 80.3, 76.5, 81.1, 78.2, 79.1],
    "B": [81.2, 86.6, 83.4, 82.9, 83.3, 77.1, 86.9, 77.9, 76.1, 83.7,
          82.3, 80.6, 81.9, 74.9, 71.9, 77.7, 77.1, 85.6, 79.8, 72.0,
          86.9, 79.7, 79.7, 79.1, 83.0, 84.6, 75.5, 81.7, 80.7, 83.6]
})

# 여기에 코드를 작성하시오.

from scipy import stats


# 독립표본 t검정
t1, p1 = stats.ttest_ind(df['A'], df['B'], alternative='two-sided', equal_var=True) # 귀무가설 채택, 유의미한 효과 x
ans1 = round(p1,3)
print("ans1: ", ans1)

# 대응표본 t검정
t2, p2 =stats.ttest_rel(df['A'], df['B']) # 귀무가설 기각, 유의미한 효과 o
ans2 = round(p2,3)
print("ans2: ", ans2)

t3, p3 =stats.wilcoxon(df['A'], df['B'])
ans3 = round(p3,3)
print("ans3: ",ans3)

```

# 1.풀이

```python
from scipy import stats

# ① 독립표본 t-검정 (양측)
t1, p1 = stats.ttest_ind(df["A"], df["B"], equal_var=True)
print("① p-value(two-sided):", round(p1, 3))

# ② 대응표본 t-검정 (양측)
t2, p2 = stats.ttest_rel(df["A"],df["B"])
print("② p-value(two-sided):", round(p2, 3))

# ③ 윌콕슨 부호순위 검정 (양측)
w3, p3 = stats.wilcoxon(df["A"], df["B"], alternative="two-sided")
print("③ p-value(wilcoxon):", round(p3, 3))
```

문제2) 제공된 데이터는 러닝 클래스 참가자와 미참가자의 10km 러닝 기록에 대한 자료이며, 두 집단 간 기록에 차이가 있는지 검정하려고 한다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오.

[제공 데이터]
1. non_part : 러닝클래스 미참가자의 10km 기록(분)
2. part : 러닝클래스 참가자의 10km 기록(분)

[문제]
1. 두 집단 각각에 대해 샤피로-윌크 검정을 수행하여 정규성을 확인하고, 두 집단 중 p-value가 가장 작은 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. 두 집단의 등분산성을 만족하는지 알아보기 위해 레빈의 검정을 수행하고 p-value 값을 구하여라.
☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. 위 ①, ②에서 구한 결과를 근거로 독립표본 t-검정, Welch’s t-검정, 만-휘트니 U 검정 중 적절한 검정을 선택하여 수행하고 p-value 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.DataFrame({
    "non_part": [49.1, 51.0, 50.0, 52.3, 50.4, 51.5, 47.7, 47.2, 49.9, 47.5,
                 49.7, 52.2, 52.3, 49.9, 48.7, 50.5, 45.1, 48.8, 53.4, 51.5],
    "part": [53.0, 47.0, 49.5, 50.8, 53.0, 54.5, 52.1, 56.0, 50.0, 44.6,
             58.3, 43.7, 49.1, 56.5, 50.0, 49.3, 51.1, 59.4, 54.3, 48.3]
})

# 여기에 코드를 작성하시오.

from scipy import stats

t1, p1 = stats.shapiro(df['non_part'])
t2, p2 =stats.shapiro(df['part'])  # 둘 다 정규성 만족
if p1<p2: ans1=round(p1,3)
else: ans1 = round(p2,3)

print("ans1:", ans1)

t3, p3 = stats.levene(df['non_part'], df['part']) # 등분산 불만족
ans2 = round(p3, 3)
print("ans2: ", ans2)

# 독립표본 t검정... 정규성 불만족 시 만휘트니유, 등분산 불만족 시 웰치스

t4, p4 = stats.ttest_ind(df['non_part'], df['part'], equal_var=False)
ans3 = round(p4, 3)
print("ans3: ", ans3)

```

# 2.풀이

```python
from scipy import stats

# ① 정규성 검정
p1 = min(stats.shapiro(df['non_part']).pvalue, stats.shapiro(df['part']).pvalue)

# ② 등분산성 검정
p2 = stats.levene(df['non_part'], df['part']).pvalue

# ③ 독립표본 t-검정 (Welch's 검정)
p3 = stats.ttest_ind(df['non_part'], df['part'], equal_var=False).pvalue

print("① 정규성 p-value :", round(p1, 3))
print("② 등분산성 p-value:", round(p2, 3))
print("③ 최종 검정 p-value:", round(p3, 3))
```
