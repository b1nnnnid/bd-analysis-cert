문제1) 제공된 데이터는 두 지역(A, B)의 10km 배달 소요시간(분)에 대한 자료이다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오.

[제공 데이터]
1. A : A지역 10km 배달 소요시간(분)
2. B : B지역 10km 배달 소요시간(분)

[문제]
1. 두 집단 각각에 대해 샤피로-윌크 검정을 수행하여 정규성을 확인하고, 두 집단 중 p-값이 가장 큰 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. 두 집단의 로그 배달 소요시간(log A, log B)에 대한 F-검정을 수행할 때 검정통계량의 값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. 두 집단의 로그 배달 소요시간의 평균 차이에 대해 독립표본 t-검정을 수행하고 p-값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.DataFrame({
    "A": [77.5, 100.8, 71.5, 104.9, 82.9, 75.4, 63.1, 80.4, 74.0, 74.9,
          91.5, 86.1, 80.5, 74.5, 68.2, 76.8, 72.3, 71.5, 119.0, 81.5,
          61.0, 77.9, 94.6, 85.1, 71.2, 78.5, 64.8, 77.6, 78.7, 67.4],
    "B": [76.5, 71.9, 70.5, 70.4, 77.8, 96.1, 89.1, 87.9, 79.1, 72.1,
          88.9, 74.9, 94.0, 68.4, 91.8, 73.3, 97.2, 90.8, 71.4, 75.2,
          68.2, 107.8, 71.4, 70.7, 84.6, 93.6, 72.4, 77.3, 87.5, 63.7]
})

# 여기에 코드를 작성하시오.

from scipy import stats

t1, p1 = stats.shapiro(df['A'])
t2, p2 = stats.shapiro(df['B']) 
ans1 = round(max(p1,p2),3)
print("ans1: ", ans1)

import numpy as np
# 로그 처리는 np.log()
var1 = np.log(df['A']).var()
var2 = np.log(df['B']).var()
if var1 > var2:
      ans2 = round(var1/var2,3)
else:
      ans2 = round(var2/var1,3)
print("ans2: ", ans2)

t3, p3 = stats.ttest_ind(np.log(df['A']), np.log(df['B']))
ans3 = round(p3, 3)
print("ans3: ", ans3)

```

# 1.풀이

```python
import numpy as np
from scipy import stats

# ① 정규성 (원자료)
pA = stats.shapiro(df["A"]).pvalue
pB = stats.shapiro(df["B"]).pvalue
print("① Shapiro (max(A, B)):", round(max(pA, pB), 3))

# ② 로그 변환 후 F-검정 통계량
logA = np.log(df["A"])
logB = np.log(df["B"])

varA = np.var(logA, ddof=1)
varB = np.var(logB, ddof=1)
F_stat = max(varA, varB) / min(varA, varB)   # 양측 F-검정용 통계량
print("② F-statistic (logA vs logB):", round(F_stat, 3))

# ③ 로그 기준 독립표본 t-검정
t_stat, p_t = stats.ttest_ind(logA, logB)
print("③ p-value (log):", round(p_t, 3))
```

문제2) 제공된 데이터는 세 지역(A, B, C)의 매장 대기시간(분)에 대한 자료이다. 소문항별로 답을 구한 후, 구한 답을 제시된 [제출 형식]에 맞춰 답안 제출 페이지에 입력하시오.

[제공 데이터]
1. A : A지역 매장 대기시간(분)
2. B : B지역 매장 대기시간(분)
3. C : C지역 매장 대기시간(분)

[문제]
1. 세 지역의 매장 대기시간이 등분산성을 만족하는지 알아보기 위해 바틀렛 검정을 수행할 때 p-값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. 세 지역의 매장 대기시간이 등분산성을 만족하는지 알아보기 위해 레빈의 검정을 수행할 때 p-값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
3. 세 지역의 매장 대기시간에 유의미한 차이가 있는지 일원분산분석을 수행하고 p-값을 구하여라.
 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

df = pd.DataFrame({
    "A": [50, 51, 49, 52, 48, 50, 51, 49, 50, 52, 48, 54],
    "B": [52, 53, 54, 51, 55, 52, 53, 54, 52, 53, 51, 65],
    "C": [54, 55, 56, 53, 55, 54, 56, 55, 54, 55, 53, 53]
})

# 여기에 코드를 작성하시오.

from scipy import stats

# 정규성 가정, 이상치 민감
t1, p1 = stats.bartlett(df['A'],df['B'],df['C'])
ans1 = round(p1, 3)
print("ans1: ", ans1) # 반올림한 소수 셋째짜리가 0이면 0.0으로 표기됨...혼동 주의 

# 이상치에 둔감
t2, p2 = stats.levene(df['A'], df['B'], df['C'])
ans2 = round(p2,3)
print("ans2: ", ans2)

# 아노바...일원분산분석
t3, p3 = stats.f_oneway(df['A'], df['B'], df['C'])
ans3 =  round(p3,3)
print("ans3: ", ans3)
```

# 2.풀이

```python
from scipy import stats
import numpy as np

# ① Bartlett (정규성 가정, 이상치에 민감) # "B" : 65 이상치
p_bart = stats.bartlett(df["A"], df["B"], df["C"]).pvalue
print("① Bartlett p-value:", round(p_bart, 3))

# ② Levene (이상치에 둔감) # "B" : 65 이상치
p_lev = stats.levene(df["A"], df["B"], df["C"]).pvalue
print("② Levene p-value:", round(p_lev, 3))

# ③ 일원분산분석 (f_oneway)
f_stat, p_anova = stats.f_oneway(df["A"], df["B"], df["C"])
print("③ ANOVA p-value:", round(p_anova, 3))
```
