 ☞ [제출 형식] 반올림하여 소수 셋째 자리까지 작성
2. 학과(컴퓨터, 수학, 경영)와 동아리 가입여부(가입/미가입) 간의 독립성 검정을 실시하였을 때, 카이제곱 통계량을 구하여라.

3. ②의 검정에서 구한 기대도수를 이용하여, 컴퓨터학과 학생이 동아리에 가입할 것으로 기대되는 수(기대빈도)를 구하여라.

# 풀이

```python
from scipy import stats

# ---------- ① 적합도 검정 (가입자만, A/B/C/D 비율 동일 가정) ----------
join_only = df[df["동아리"] != "미가입"]["동아리"]
obs = join_only.value_counts().values
chi2_gof, p_gof = stats.chisquare(obs)
print("① GOF p-value:", round(p_gof, 3))

# ---------- ② 독립성 검정 (학과 × 가입여부) ----------
def classify(row):
  if row["동아리"] == "미가입":
    return False
  else:
    return True

df['가입여부'] = df.apply(classify, axis=1)
tab = pd.crosstab(df["학과"], df["가입여부"])
chi2_ind, p_ind, dof_ind, exp_ind = stats.chi2_contingency(tab)
print("② 독립성 χ²:", round(chi2_ind, 3))

# ---------- ③ 기대빈도(컴퓨터 × 가입여부) ----------
exp_table = pd.DataFrame(exp_ind, index=tab.index, columns=tab.columns)
expected_comp_join = exp_table.loc["컴퓨터", True]
print("③ 기대도수(컴퓨터, 가입여부):", int(round(expected_comp_join)))
```
