**독립표본 t-검정**
- 사례) 두 반(반 A, 반 B) 학생들의 평균 시험 점수가 같은지 검정하고자 한다. 각 반에서 무작위로 30명씩 추출하여 점수를 측정하였다.



```python
# 독립표본 t검정: 두 집단의 평균 비교, 표본간 독립...ttest_ind()


import pandas as pd

df = pd.DataFrame({
    "A": [79, 81, 78, 84, 80, 85, 77, 82, 83, 79,
          80, 84, 82, 76, 85, 81, 79, 83, 78, 82,
          80, 86, 77, 84, 81, 79, 82, 80, 83, 78],
    "B": [83, 84, 82, 87, 86, 83, 80, 85, 86, 81,
          82, 88, 85, 80, 86, 84, 83, 86, 82, 85,
          84, 87, 81, 86, 85, 82, 84, 83, 86, 82]
})
```

1. 통계적 가설 설정
 - 귀무가설(H0) : 𝝁_𝑨= 𝝁_𝑩 (두 반의 평균 점수는 같다)
 - 대립가설(H1) : 𝝁_𝑨≠ 𝝁_𝑩 (두 반의 평균 점수는 다르다) → 양측검정
2. 유의수준 결정
 - 유의수준 𝜶=𝟎.𝟎𝟓

3. 양측검정, 등분산성 만족 (기본값 alternative=‘two-sided’, equal_var=True)
4. 결과도출
 - t-statistic과 p-value를 확인하고, p-value < 0.05이면 귀무가설 기각

```python
# 독립표본 t검정은 등분산성을 가정... equal_var=True가 기본값

from scipy import stats

t_stat, p_two = stats.ttest_ind(df["A"], df["B"])
print("t-statistic:", t_stat)
print("p-value(two-sided):", p_two)
```

 -  해석 : 유의수준 0.05에서 p-value가 2.15e-05이므로, 귀무가설을 기각 → 두 반 평균은 통계적으로 다르다.

5. 단측검정(<)
 - 귀무가설(H0) : 𝝁_𝑨= 𝝁_𝑩
 - 대립가설(H1) : 𝝁_𝑨< 𝝁_𝑩(B반의 평균 점수가 더 높다)


```python
# t검정은 모수 검정(정규성 만족)을 전제... 불만족이면 비모수검정 >> 만휘트니 u 검정정

stats.ttest_ind(df["A"], df["B"], alternative="less")
```

 - 해석 : 유의수준 0.05에서 p-value가 5.75e-06이므로, 귀무가설을 기각 → B반의 평균 점수가 더 높다.


6. 비모수 대안
 - 정규성 가정이 불만족이라면, 만휘트니 U검정을 수행한다.

```python
stats.shapiro(df['A']), stats.shapiro(df['B'])
```

※ 만약, 위의 결과에서 정규성이 하나라도 만족하지 않으면 Mann-Whitney U 검정 수행
- 지금은 만족하므로 독립표본 t-검정을 수행하면 되며, 아래 코드는 참고용

```python
stats.mannwhitneyu(df["A"], df["B"], alternative="two-sided")
```

**등분산 검정**

```python
# 레빈의 검정(H0: 두 집단은 등분산을 만족한다)

stats.levene(df["A"], df["B"])
```

```python
# 등분산 만족하니까 독립표본 t검정 수행

stats.ttest_ind(df["A"], df["B"])
```

**Welch의 t-검정**

```python
# 등분산 불만족시 welch's t검정 수행
# equal_var=False만 해주면됨
stats.ttest_ind(df["A"], df["B"],  equal_var=False)
```

**대응표본 t-검정**
- 사례) 같은 학생 24명을 대상으로 보강 수업 전/후 점수를 비교한다. 평균 향상이 있는지 검정하라.



```python
# 같은 집단의 사전/사후 평균 비교, 표본 간 대응 관계...ttest_rel()

import pandas as pd

df_paired = pd.DataFrame({
    "before": [72, 75, 78, 70, 76, 79, 74, 77, 73, 75, 76, 72,
               74, 77, 71, 78, 73, 76, 75, 72, 74, 76, 73, 75],
    "after":  [74, 77, 80, 72, 78, 80, 76, 79, 74, 77, 78, 73,
               76, 79, 72, 80, 74, 78, 77, 73, 76, 78, 74, 77]
})
```

1. 통계적 가설 설정
 - 귀무가설(H0) : 𝝁_𝒅𝒊𝒇𝒇= 0 (전후 평균 차이가 없다)
 - 대립가설(H1) : 𝝁_𝒅𝒊𝒇𝒇≠ 0 (전후 평균 차이가 있다) → 양측검정

2. 유의수준 결정
 - 유의수준 𝜶=𝟎.𝟎𝟓

3. 양측검정 만족 (기본값 alternative=‘two-sided’)
4. 결과도출
 - t-statistic과 p-value를 확인하고, p-value < 0.05이면 귀무가설 기각

```python
t_stat, p_two = stats.ttest_rel(df_paired["after"], df_paired["before"])
print("t-statistic:", t_stat)
print("p-value(two-sided):", p_two)
```

 -  해석 : 유의수준 0.05에서 p-value가 4.60e-15이므로, 귀무가설을 기각 → 보강 수업 후 점수가 유의하게 변화하였다.

5. 단측검정(>)
 - 귀무가설(H0) : 𝝁_𝒅𝒊𝒇𝒇= 0 (전후 평균 차이가 없다)
 -  대립가설(H1) : 𝝁_𝒅𝒊𝒇𝒇> 0 (전후 평균이 증가하였다)

```python
stats.ttest_rel(df_paired["after"], df_paired["before"], alternative="greater")
```

 -  해석 : 유의수준 0.05에서 p-value가 2.30e-15이므로, 귀무가설을 기각 → 전후 평균이 증가하였다.


6. 비모수 대안

```python
stats.shapiro(df_paired['after']-df_paired['before']) # 정규성 검정... 대응표본은 둘의 차이에 대해, 정규분포 x
```

```python
# 정규성 안 따르면 윌콕슨 부호순위 검정

stats.wilcoxon(df_paired['after'], df_paired['before'], alternative='two-sided')
```
