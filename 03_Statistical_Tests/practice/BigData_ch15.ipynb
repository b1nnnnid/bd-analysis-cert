**단일표본 t-검정**
- 사례) A 공장에서 생산하는 제품의 평균 길이가 150cm인지 알아보고자 한다.
이를 위해 무작위로 30개의 제품 길이(cm)를 측정한 결과는 다음과 같다.


```python
# 가설검정: 모집단의 특성에 대한 주장을 가설로 세우고 표본조사로 가설의 채택여부를 결정하는 과정
# 귀무가설(H0, 일반이론)/대립가설(H1, 증명하려는 이론)

# 유의수준 알파: 귀무가설이 참일 때 1종 오류를 범할 확률의 허용 한계(보통 0.05)
# 유의확률 p-value: 귀무가설을 지지하는 정도를 나타내는 확률
# 알파가 p보다 작으면 기각, 크면 채택

# 검정통계량: 귀무가설의 참거짓 판단으 위해 사용하는 통계적 계산값
# 기각역/채택역
# 양측검정/단측검정 


# 단일표본 t 검정: 모평균과 특정값 사이 차이 검정 ...ttest_1samp
import pandas as pd

df = pd.DataFrame({
    'product': [153, 152, 154, 151, 155, 152, 153, 151, 154, 152,
                153, 152, 154, 153, 152, 151, 155, 154, 152, 153,
                152, 153, 151, 154, 152, 153, 152, 155, 154, 152]
})
```

1. 통계적 가설 설정
 - 귀무가설(H0) : 𝝁=𝟏𝟓𝟎 (제품 평균 길이는 150cm 이다)
 - 대립가설(H1) : 𝝁≠𝟏𝟓𝟎 (제품 평균 길이는 150cm가 아니다) → 양측검정
2. 유의수준 결정
 - 유의수준 𝜶=𝟎.𝟎𝟓


3. 양측검정 (기본값 alternative=‘two-sided’)
4. 결과도출
 - t-statistic과 p-value를 확인하고, p-value < 0.05이면 귀무가설 기각

```python
from scipy import stats

t_stat, p_value = stats.ttest_1samp(df['product'], 150)
print("t-statistic:", t_stat) # 검정통계량
print("p-value:", p_value) # 유의확률
# p-value가 더 작으니까 기각
```

 -  해석 : 유의수준 0.05에서 p-value가 2.61e-13이므로, 귀무가설을 기각 → 평균이 150cm라 볼 수 없다.

5. 단측검정(>)
 - 귀무가설(H0) : 𝝁=𝟏𝟓𝟎
 - 대립가설(H1) : 𝝁>𝟏𝟓𝟎 (제품 평균 길이는 150cm보다 크다)

```python
# alternative = 'greater' or 'less'

stats.ttest_1samp(df['product'], 150, alternative='greater')
```

 - 해석 : 유의수준 0.05에서 p-value가 1.30e-13이므로, 귀무가설을 기각 → 평균이 150cm보다 크다.


6. 단측검정(<)
 - 귀무가설(H0) : 𝝁=𝟏𝟓𝟎
 - 대립가설(H1) : 𝝁<𝟏𝟓𝟎 (제품 평균 길이는 150cm보다 작다)


```python
stats.ttest_1samp(df['product'], 150, alternative='less')
```

 - 해석 : 유의수준 0.05에서 p-value가 0.99이므로, 귀무가설을 채택


**정규성 검정**

```python
# t검정 활용 위해서는 표본 데이터가 정규분포를 따르는지 확인 필요
# 샤피로 윌크 통해 정규성 판단(H0: 데이터는 정규분포를 따른다)...모수검정

stats.shapiro(df['product'])
```

**비모수 검정**

```python
# 데이터가 정규분포를 따르지 않을 때, 윌콕슨 부호순위 검정 활용(중앙값 차이를 검정)
# 귀무가설: 중앙값 =150
# 대립가설: 중앙값 != 150

stats.wilcoxon(df['product'] - 150, alternative='two-sided')
```
