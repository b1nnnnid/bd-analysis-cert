# Ch06. 인덱스 선택

```python
import pandas as pd
df = pd.read_csv('coin.csv')
df

# pandas 기본 인덱스: 0,1,2 ... 숫자
```

```python
# Date를 인덱스로 설정(df= df~~ 해서 넣어줘야 원본 데이터에 반영)

df = df.set_index('Date')
# ~~~read_csv(~~~, index_col=0)
df
```

```python
print("-----특정 인덱스----")
print(df.loc['2018-01-01'])         # 인덱스가 날짜인 경우

print("-----인덱스 범위----")
print(df.loc[:, ['Open']])          # 열 선택

print("-----특정 행, 열----")
print(df.loc['2018-01-01', 'Open']) # 특정 행, 열
```

```python
# 기존 loc는 라벨명 설정
# iloc는 숫자 인덱스로

print("-----특정 인덱스----")
print(df.iloc[0])            # 0 번째 행

print("-----인덱스 범위----")
print(df.iloc[:, [1, 2]])      # 1 번째, 2 번째 열

print("-----특정 행, 열----")
print(df.iloc[0, 1])           # 0 번째 행, 1 번째 열
```

```python
data = {'이름' : ['영희', '철수'],
        '나이' : [20, 22],
        '성별' : ['여', '남'],
        '점수' : [95, 90]}
df = pd.DataFrame(data)
df
```

```python
# inplace=True이면 원본데이터를 직접 변경(앞에 df= 으로 넣어주지 않아도)

df.set_index('이름', inplace=True)
df
```

```python
print(df.loc['영희',['성별','점수']])
print(df.iloc[0, [1, 2]])
print(df.loc['영희', '성별' : '점수'])
print(df.iloc[0, 1:])
```

```python
data = {'이름' : ['영희', '철수'],
        '나이' : [20, 22],
        '성별' : ['여', '남'],
        '점수' : [95, 90]}
df = pd.DataFrame(data)
df.set_index('이름', inplace=True)
df
```

```python
#1. loc : 슬라이싱 끝 포함
print(df.loc['영희', '성별' : '점수'])

#2. iloc : 슬라이싱 끝 미포함
print(df.iloc[0, 1:2])
```
