#Ch07 연습문제

# 1.다음 코드를 실행했을 때 출력 결과를 예측하시오.

```python
import pandas as pd

data = {
    '이름': ['영희', '철수'],
    '나이': [20, 22],
    '점수': [95, 90]
}
df = pd.DataFrame(data)
df.set_index('이름', inplace=True)

df.loc['영희', '점수'] = 100
print(df)
```

# 2.다음 코드를 실행했을 때 출력 결과를 예측하시오.

```python
df = df.drop('나이', axis=1)
print(df)
```

# 3.다음 코드를 실행했을 때 출력 결과를 예측하고, inplace=True를 사용했을 때의 차이를 간단히 설명하시오.

```python
new_df = df.drop('점수', axis=1)
print(new_df)
print(df)
```

- inplace=False가 기본이므로 원본 df는 변하지 않고, 삭제된 새 DataFrame(new_df)이 반환된다.

- inplace=True를 사용하면 df가 직접 변경되며, drop()의 반환값은 None이 된다.
