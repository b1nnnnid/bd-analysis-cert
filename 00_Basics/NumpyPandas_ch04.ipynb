```python
weather = pd.read_csv('weather.csv')
weather.describe()
# лҚ°мқҙн„°мқҳ мЈјмҡ” нҶөкі„ мҡ”м•Ҫ м •ліҙ нҷ•мқё
```

```python
# м§‘кі„ н•ЁмҲҳ

# 1. mean() - нҸүк· мқҳ л°ҳнҷҳ
print(weather.mean())
# 2. std() - н‘ңмӨҖнҺём°Ёмқҳ л°ҳнҷҳ
print(weather.std())
```

```python
import numpy as np
```

```python
# л‘җ н‘ңмӨҖнҺём°Ёмқҳ м°Ёмқҙ
# ddof : delta degree of freedom(мһҗмң лҸ„ м–јл§ҲлӮҳ)

# 1. л„ҳнҢҢмқҙ н‘ңмӨҖнҺём°Ё
data = [1, 2, 3, 4, 5]
print("Numpy лӘЁм§‘лӢЁ н‘ңмӨҖнҺём°Ё:", np.std(data))         # кё°ліём ҒмңјлЎң лӘЁм§‘лӢЁ н‘ңмӨҖнҺём°Ё(ddof=0)... nмңјлЎң лӮҳлҲҲ к°’
print("Numpy н‘ңліё н‘ңмӨҖнҺём°Ё:", np.std(data, ddof=1))   # ddof=1лЎң н‘ңліё н‘ңмӨҖнҺём°Ё

# 2. нҢҗлӢӨмҠӨ н‘ңмӨҖнҺём°Ё
df = pd.DataFrame({'values': [1, 2, 3, 4, 5]})
print("Pandas лӘЁм§‘лӢЁ н‘ңмӨҖнҺём°Ё:", df['values'].std(ddof=0))  # ddof=0мңјлЎң лӘЁм§‘лӢЁ н‘ңмӨҖнҺём°Ё
print("Pandas н‘ңліё н‘ңмӨҖнҺём°Ё:", df['values'].std())          # кё°ліём ҒмңјлЎң н‘ңліё н‘ңмӨҖнҺём°Ё(ddof=1)... n-1лЎң лӮҳлҲҲ к°’
```
