# Ch08. 데이터 필터링

```python
import pandas as pd
data = {
    "이름" : ["철수", "영희", "민수", "지은"],
    "국어" : [90, 80, 85, 100],
    "수학" : [85, 95, 88, 100],
    "영어" : [80, 100, 75, 90]
}
df = pd.DataFrame(data)
df.set_index("이름", inplace=True)
df
```

```python
# Boolean 인덱싱: 조건식 < > == != 등을 적용해 true/false 반환

df['국어'] > 85
```

```python
df[df['국어'] > 85]
```

```python
cond1 = df['국어'] > 85
cond2 = df['영어'] > 85
```

```python
# 조건 조합: 합집합/교집합을 활용해 2개 이상의 조건 조합
# 교집합 &: 둘 다 참이어야 참 

df[cond1 & cond2]
```

```python
# 합집합 |: 하나만 참이어도 참

df[cond1 | cond2]
```

```python
# 특정 목록에 포함되는 값만 필터링

cond = df['수학'].isin([100, 95])
df[cond]
```

```python
score = [100, 95]
cond = df['수학'].isin(score)
df[cond]
```
