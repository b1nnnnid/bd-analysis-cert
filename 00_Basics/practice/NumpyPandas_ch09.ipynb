# Ch09. DataFrame 병합

```python
import pandas as pd

df1 = pd.DataFrame({'이름': ['영희', '철수'], '점수': [90, 85]})
df2 = pd.DataFrame({'이름': ['민수', '지영'], '점수': [88, 92]})

print(df1)
print(df2)
```

```python
# concat 함수: 데이터프레임 병합(행 병합은 axis=0, 열 병합은 axis=1)
# axis 옵션은 항상 0이 기본값

result1 = pd.concat([df1, df2])
print(result1)
```

```python
df3 = pd.DataFrame({'성별': ['여', '남']})
result = pd.concat([df1, df3], axis=1)
print(result)
```

```python
# 병합 후 인덱스를 0부터 재설정

pd.concat([df1, df2], ignore_index=True)
```

```python
# keys : 계층적 인덱스 추가

pd.concat([df1, df2], keys=['상반기', '하반기'])
```
