# Ch05. DataFrame 열/행 추가

```python
import pandas as pd
df = pd.read_csv('countries.csv', index_col=0)
# 0번째 칼럼(국가 약자)을 인덱스로 지정
df
```

```python
df['density']=0
# 새로운 칼럼 생성
df
```

```python
df2 = pd.read_csv('countries.csv', index_col=0)

# 열에 통째로 접근해 여러 행을 한 번에 계산 -> 인구밀도 = 인구/면적
df2['density'] = df2['population'] / df2['area']
df2
```

```python
data = {'이름' : ['영희', '철수'],
        '나이' : [20, 22],
        '성별' : ['여', '남'],
        '점수' : [95, 90]}
df = pd.DataFrame(data)
df
```

```python
# 새로운 행 추가 : loc(location)... 2번 행 추가
df.loc[2] = 0
df
```

```python
# 각 칼럼마다 값을 리스트로 부여
# 기존에 값 있을 경우 덮어쓰기
df.loc[3] = ['민호', 21, '남', 50]
df
```

```python
# 백업 라벨로 행 추가(행 추가 시 인덱스명 지정 가능)

df.loc['백업'] = df.loc[1]
df
```
