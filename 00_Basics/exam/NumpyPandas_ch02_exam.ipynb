## 1. 다음 코드를 실행했을 때 출력될 값을 에측하시오.

```python
import numpy as np
a = np.array([[1, 2], [3, 4], [5, 6]])
print(a.ndim, a.shape, a.size, a.dtype)
# 차원, 모양, 크기, 자료형
```

## 2. a.reshape(3, 2)과 a.transpose()의 차이점에 대해 배열 a = np.array([1, 2, 3], [4, 5, 6]])를 기준으로 설명하시오.

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a.reshape(3, 2))
print(a.transpose())
# 모양 바꾸기, 전치
```

## 3. 다음 코드의 실행 결과는?

```python
a = np.array([3, 3, 2, 1, 2, 1])
print(np.unique(a))
# 중복값 제거 후 정렬
```
