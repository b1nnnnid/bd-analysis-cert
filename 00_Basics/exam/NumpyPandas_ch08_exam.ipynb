# 1.다음 코드를 실행했을 때 출력 결과를 예측하시오.

```python
import pandas as pd

data = {
    '이름': ['영희', '철수', '민수', '지영'],
    '점수': [92, 87, 85, 79],
    '성별': ['여', '남', '남', '여']
}
df = pd.DataFrame(data)

print(df[df['점수'] == 85])
```

# 2.다음 코드를 실행했을 때 출력 결과를 예측하시오. (힌트: ~은 NOT연산자로서 원래 조건의 반대를 출력한다.)

```python
print(df[~df['성별'].isin(['남'])])
```

# 3.다음 코드를 실행했을 때 출력 결과를 예측하시오.


```python
print(df[(df['점수'] > 85) & (df['성별'] == '여')])
```

```python
print(df[df['이름'].isin(['민수', '지영'])])
```
