# 1.다음 코드를 실행했을 때 출력 결과를 예측하시오.

```python
import pandas as pd
data = {
    '이름': ['영희', '철수'],
    '나이': [20, 22],
    '성별': ['여', '남'],
    '점수': [95, 90]
}
df = pd.DataFrame(data)
df.set_index('이름', inplace=True)
print(df.iloc[0, 1:2])
print(df.iloc[0, 1:])
```
