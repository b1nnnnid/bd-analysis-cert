#1. 다음 코드의 출력 결과를 예측하시오.

#2. 다음 코드의 실행 결과를 예측하시오. (힌트 : 시리즈(Series)는 인덱스를 통해 값에 접근할 수 있다.)

```python
import pandas as pd
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
# 1번 인덱스 접근
print(s.iloc[1])
print(s['b'])
```

#3. 다음 코드의 실행 결과를 예측하고, 결측값 처리 방법에 대해 간단히 설명하시오. (힌트 : 데이터프레임의 fillna 함수는 결측값을 대체하는 함수이다.)

```python
import pandas as pd
import numpy as np
data = {'이름': ['철수', '영희', '민수'], '성적': [np.nan, 90, 85]}
df = pd.DataFrame(data)
print(df.fillna(0))
```
