# Ch05. (1유형) 정렬

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve',
             'Frank', 'Grace', 'Hannah', 'Ian', 'Judy'],
    'Salary': [3200, 3400, 3100, 8000, 3300,
               3600, 9000, 3000, 3400, 9500],
    'Age': [25, 28, 26, 27, 29,
            30, 26, 31, 32, 28]
}
df = pd.DataFrame(data)
df
```

```python
# 단일 컬럼 정렬 sort_values
# by: 정렬 기준이 되는 컬럼명, 디폴트는 오름차순(ascending = True)

df.sort_values(by='Salary')
```

```python
# 내림차순
df.sort_values(by='Age', ascending=False)
```

```python
# 다중 컬럼 기준 정렬
# 연봉 오름차순, 그 중에 같은 값이 있으면 그때 나이로 내림차순
df.sort_values(by=['Salary', 'Age'], ascending=[True, False])
```

```python
df = pd.DataFrame({'A': [3, 1, 2]})

# inplace=False (기본값)
new_df = df.sort_values(by='A')   # 정렬된 새 객체 반환
print(new_df)
print(df)                         # 원본은 그대로 유지

# inplace=True
df.sort_values(by='A', inplace=True)  # 원본 df 자체가 정렬됨
print(df)                             # 원본이 바뀜, 메서드는 None 반환
```
