# Ch07. (1유형) 그룹핑

```python
# 그룹핑(Groupby): 데이터를 특정 기준으로 묶어 요약 통계를 계산하는 과정
# 그룹별 평균, 합계, 개수 등을 계산하여 집단 간 비교 가능
# df.groupby('기준 컬럼')['계산 컬럼'].집계함수()

import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Salary': [3000, 3500, 2800, 2900, 4000, 4200]
}

df = pd.DataFrame(data)

# 단일 그룹핑... groupby는 분할 -> 처리 -> 병합 순으로 작동

# 부서별 평균 연봉
grouped = df.groupby('Department')['Salary'].mean()
grouped
```

```python
data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Salary': [3000, 3500, 2800, 2900, 4000, 4200]
}

df = pd.DataFrame(data)

# 여러 기준으로 그룹핑: 기준 컬럼 자리에 컬럼을 리스트로 전달

# 부서별 + 성별 평균 연봉
grouped = df.groupby(['Department', 'Gender'])['Salary'].mean()
grouped
```

```python
import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Salary': [3000, 3500, 2800, 2900, 4000, 4200],
    'Bonus': [200, 250, 180, 160, 300, 320]
}

df = pd.DataFrame(data)

# 부서별 평균 급여와 총 보너스를 함께 계산
# agg(): 여러 집계함수 적용(딕셔너리 형태)
result = df.groupby('Department').agg({'Salary': 'mean', 'Bonus': 'sum'})
result
```

```python
data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Salary': [3000, None, 2800, 2900, 4000, 4200]
}
df = pd.DataFrame(data)

# transform('mean'): 각 그룹의 평균값이 원래 행에 그대로 반영
# 행 수가 유지 -> 결측값 채우기에 적합

df.groupby('Department')['Salary'].transform('mean')
```

```python
# 연봉에 결측값 -> 해당 부서의 연봉 평균으로 대체

df['Salary'] = df['Salary'].fillna(df.groupby('Department')['Salary'].transform('mean'))
df
```

```python
grouped = df.groupby('Department')['Salary'].mean()
grouped
```

```python
# idxmax(): 가장 큰 값을 가진 인덱스(그룹명 반환)
# idxmin(): 가장 작은 값을 가진 그룹명 반환

print(grouped.idxmax())  # 평균 급여 가장 높은 부서
print(grouped.idxmin())  # 평균 급여 가장 낮은 부서
```
