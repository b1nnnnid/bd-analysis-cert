# Ch09. (1유형) 문자열 처리

```python
import pandas as pd

data = {
    'Email': ['kim@gmail.com', 'lee@naver.com', 'park@daum.net', 'choi@gmail.com'],
    'Name': ['Kim', 'Lee', 'Park', 'Choi'],
    'Comment': ['Hello!!', 'Nice to meet you :)', '  Thanks!! ', 'ok~~~']
}

df = pd.DataFrame(data)
df

# 텍스트 컬럼의 정제, 분류, 검색 등 전처리 과정에서 문자열 처리 함수가 자주 사용
```

```python
# split(): 문자열 분리
# 앞에 str. 빼먹지 말기

df['Email'].str.split('@').str[1]
```

```python
# lower(): 소문자 변환
# upper(): 대문자 변환

df['Name'].str.lower()
```

```python
# 2번 인덱스 접근

df['Comment'].loc[2]
```

```python
# strip(): 양측 공백 제거
# lstrip()/rstrip(): 좌측/우측 공백 제거

df['Comment'].str.strip().loc[2]
```

```python
# contains(): 특정 문자열 포함 여부...Boolean 반환 -> df로 필터링

df[df['Email'].str.contains('gmail')]
```

```python
# replace(): 문자열 치환
# regex =True: 정규표현식으로 패턴 인식

df['Comment'].str.replace('!', '', regex=False)
```

```python
# 문자열 슬라이싱

df['Email'].str[:5]
```
