# Ch08. (1유형) 피벗테이블

```python
import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'Sales', 'IT', 'IT'],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'M', 'F'],
    'Salary': [3000, 3200, 2500, 2700, 3100, 4000, 3900]
}

df = pd.DataFrame(data)
df
```

```python
# 피벗테이블: 데이터를 요약(집계)하고 차원 축소하여 가독성 높은 형태로 재구성하는 도구
# 행 인덱스, 값을 지정해 집계함수(aggfunc)로 요약
# pd.pivot_table(index='column a', columns='column b', values='column c', aggfunc='mean')

pd.pivot_table(df, index='Department', values='Salary', aggfunc='mean')
```

```python
pd.pivot_table(df, index='Department', columns='Gender', values='Salary', aggfunc='mean')
```

```python
# 여러 집계 함수를 리스트로 활용

pd.pivot_table(df, index='Department', values='Salary',
               aggfunc=['mean', 'max', 'min'])
```

```python
pivot = pd.pivot_table(df, index='Department', columns='Gender',
                       values='Salary', aggfunc='mean')
pivot
```

```python
# stack(): wide -> long
# unstack(): long -> wide

pivot1 = pivot.stack()
pivot1
```

```python
pivot2 = pivot1.unstack()
pivot2
```
