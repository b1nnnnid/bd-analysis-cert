# Ch02. (1유형) 결측값 처리

```python
import pandas as pd

data = {
    '이름': ['철수', '영희', '민수', '선호'],
    '나이': [25, None, 30, 28],
    '성별': ['남', '여', '남', None]
}
df = pd.DataFrame(data)
df
```

```python
df.isnull()
#각 원소가 결측값인지를 Boolean으로 반환(null이면 true)
```

```python
df.isnull().sum()
# 열 단위로 결측값 개수 집계
#df.isnull().sum().sum()
# 총 결측값 개수 집계
```

```python
# 결측값을 원하는 값으로 채움(특정값/평균값/최빈값... 여기서는 특정값0)
df['나이'].fillna(0)
```

```python
# 나이의 평균값으로 결측값 대체
df['나이'].fillna(df['나이'].mean())
```

```python
# 나이의 평균값으로 결측값 대체
# mode함수: 최빈값을 Series 형태로 반환(최빈값이 여러 개일 수 있기 때문에)
df['성별'].fillna(df['성별'].mode()[0]) # [0]으로 인덱스 지정해줘야 함. -> 최빈값 중 첫값만 사용
```

```python
df['성별'].mode()
# 최빈값이 한 개
```

```python
# 이전 값으로 채우기 forward: 철수의 나이로
df.ffill()
```

```python
# 다음 값으로 채우기 backward: 민수의 나이로
df.bfill()
```

```python
# 결측값 제거
# dropna(): 결측값 있는 전체 행 제거
df.dropna()
# 나이 결측 영희, 성별 결측 선호 제거
```

```python
# dropna(subset=["열이름"]): 특정 열 기준으로 제거
df.dropna(subset=['성별'])
# 성별 결측 선호만 제거
```
