제공된 데이터 sales.csv는 온라인 쇼핑몰에서 발생한 판매 기록이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Date : 거래 일자
2. Product Category : 제품 카테고리
3. Units Sold : 판매 수량
4. Unit Price : 단가
5. Payment Method : 결제 수단

[수행 순서]
1. 제품 카테고리별 총 판매 금액(판매 수량 x 단가)을 계산한다.
2. 가장 높은 총 판매 금액을 기록한 제품 카테고리의 총 금액을 구한다.
3. 결제 수단별 총 거래 건수를 계산한다.
4. 가장 거래 건수가 적은 결제 수단의 건수를 구한다.

[문제]
1. 수행 순서 2)에서 계산한 값을 입력하시오.
☞ [제출 형식] 정수(integer)로 작성
2. 수행 순서 4)에서 계산한 값을 입력하시오.
☞ [제출 형식] 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv("../data/sales.csv")

df['Total Price'] = df['Units Sold']*df['Unit Price']
category_sales = df.groupby('Product Category')['Total Price'].sum()
ans1 = int(category_sales.max())

print("ans1: ", ans1)

payment_pivot = pd.pivot_table(df, index = 'Payment Method', aggfunc = 'count')
payment_sales = payment_pivot['Units Sold'].min()

ans2 = int(payment_sales)

print("ans2: ", ans2)

# 여기에 코드를 작성하시오.
```

# 풀이(#1)

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/1_DataSet/sales.csv")

# 1. 총 판매 금액(판매 수량 x 단가)을 계산한다.
df['Total Sales'] = df['Units Sold'] * df['Unit Price']

# 2. 가장 높은 총 판매 금액을 기록한 제품 카테고리의 총 금액을 구한다.
group_sales = df.groupby('Product Category')['Total Sales'].sum()
max_sales = int(group_sales.max())

# 3. 결제 수단별 총 거래 건수를 계산한다.
# 4. 가장 거래 건수가 적은 결제 수단의 건수를 구한다.
group_payment = df.groupby('Payment Method').size()
min_payment = int(group_payment.min())

# 결과
print("① 가장 높은 총 판매 금액 카테고리의 총 금액:", max_sales)
print("② 가장 거래 건수 적은 결제 수단의 건수:", min_payment)
```

# 풀이(#2)

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/1_DataSet/sales.csv")

# 1. 총 판매 금액(판매 수량 x 단가)을 계산한다.
df['Total Sales'] = df['Units Sold'] * df['Unit Price']

# 2. 가장 높은 총 판매 금액을 기록한 제품 카테고리의 총 금액을 구한다.
pivot_sales = df.pivot_table(index='Product Category', values='Total Sales', aggfunc='sum')
max_sales = int(pivot_sales['Total Sales'].max())

# 3. 결제 수단별 총 거래 건수를 계산한다.
# 4. 가장 거래 건수가 적은 결제 수단의 건수를 구한다.
pivot_payment = df.pivot_table(index='Payment Method', aggfunc='count')
min_payment = int(pivot_payment['Units Sold'].min())

# 결과
print("① 가장 높은 총 판매 금액 카테고리의 총 금액:", max_sales)
print("② 가장 거래 건수 적은 결제 수단의 건수:", min_payment)
```
