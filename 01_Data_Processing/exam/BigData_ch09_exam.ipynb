# 풀이

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/1_DataSet/hr_data.csv")

# 1. Email 컬럼에서 @ 기준으로 분리하여, 도메인 주소만 추출한 새로운 Domain 컬럼을 생성한다.
df['Domain'] = df['Email'].str.split('@').str[1]

# 2. 전체 직원 중, 도메인이 "bilearner.com"인 직원 수를 구한다.
bilearner_df = df[df['Domain'] == 'bilearner.com']
A = len(bilearner_df)

# 3. 도메인이 "bilearner.com"인 직원 중, 이름(Name)의 문자 길이가 6자 이상인 직원 수를 구한다.
bilearner_df['Name_Length'] = bilearner_df['Name'].str.len()
B = len(bilearner_df[bilearner_df['Name_Length'] >= 6])

# 결과
print("① 도메인이 bilearner.com인 직원 수:", A)
print("② 해당 직원 중 이름 길이 6자 이상인 직원 수:", B)
```
