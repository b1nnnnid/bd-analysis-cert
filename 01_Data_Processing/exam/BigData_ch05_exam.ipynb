아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. housing_median_age : 주택 중간 연식
2. total_rooms : 총 방 수
3. population : 인구 수
4. households : 세대 수
5. median_income : 중간 소득
6. median_house_value : 주택 중간 가격

[수행 순서]
1. room_per_person 파생변수를 생성한다.
※ room_per_person = total_rooms / population
2. room_per_person 기준으로 내림차순 정렬한다.
3. 정렬된 데이터에서 상위 10개 행의 median_house_value 평균을 구한다.
4. room_per_person이 전체 평균 이상인 데이터 중에서, housing_median_age가 30 이상인 행의 수를 구한다.

[문제]

# 풀이

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/1_DataSet/housing.csv')

# 1. room_per_person 파생변수를 생성한다.
# ※ room_per_person = total_rooms / population
df['room_per_person'] = df['total_rooms'] / df['population']

# 2. room_per_person 기준으로 내림차순 정렬한다.
sorted_df = df.sort_values(by='room_per_person', ascending=False)

# 3. 정렬된 데이터에서 상위 10개 행의 median_house_value 평균을 구한다.
top10_mean = int(sorted_df.head(10)['median_house_value'].mean())

# 4. room_per_person이 전체 평균 이상인 데이터 중에서, housing_median_age가 30 이상인 행의 수를 구한다.
avg_rpp = df['room_per_person'].mean()
filtered_count = df[(df['room_per_person'] >= avg_rpp) & (df['housing_median_age'] >= 30)].shape[0]

# 결과
print("① 상위 10개의 median_house_value 평균값:", top10_mean)
print("② room_per_person 평균 이상 & housing_median_age 30 이상 수:", filtered_count)
```
