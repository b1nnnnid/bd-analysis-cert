아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. housing_median_age : 주택 중간 연식
2. total_rooms : 총 방 수
3. population : 인구 수
4. households : 세대 수
5. median_income : 중간 소득
6. median_house_value : 주택 중간 가격

[수행 순서]
1. median_house_value 컬럼에서 IQR 기준으로 이상값을 판단한다.
※ 이상값 기준 : 하한선(Q1-1.5 x IQR), 상한선(Q3+1.5 x IQR)
2. 위 기준을 벗어나는 이상값들을 제거한다.
3. 이상값이 제거된 데이터에 대해 value_per_pop 파생변수를 생성한다.
※ value_per_pop = median_house_value / population
4. value_per_pop의 평균을 구한다.

[문제]
1. 수행 순서 1)에서 계산한 IQR 값을 입력하시오.
☞ [제출 형식] 정수(integer)로 작성
2. 수행 순서 4)에서 계산한 값을 입력하시오.
☞ [제출 형식] 반올림하여 소수 셋째자리까지 작성

```python
import pandas as pd

df = pd.read_csv('../data/housing.csv')

Q1 = df['median_house_value'].quantile(0.25)
Q3 = df['median_house_value'].quantile(0.75)
IQR = Q3 - Q1
print("ans1: ", int(IQR))

df_filtered = df[(df['median_house_value'] >= Q1 - 1.5*IQR) & (df['median_house_value'] <= Q3 + 1.5*IQR)]
df_filtered['value_per_pop'] = df_filtered['median_house_value']/df_filtered['population']
print("ans2: ", round(df_filtered['value_per_pop'].mean(),3))

# 여기에 코드를 작성하시오.
```

# 풀이

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/1_DataSet/housing.csv')

# 1. median_house_value 컬럼에서 IQR 기준으로 이상값을 판단한다.
# ※ 이상값 기준 : 하한선(Q1-1.5 x IQR), 상한선(Q3+1.5 x IQR)
q1 = df['median_house_value'].quantile(.25)
q3 = df['median_house_value'].quantile(.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# 2. 위 기준을 벗어나는 이상값들을 제거한다.
df_filtered = df[(df['median_house_value'] >= lower_bound) & (df['median_house_value'] <= upper_bound)].copy()

# 3. 이상값이 제거된 데이터에 대해 value_per_pop 파생변수를 생성한다.
# ※ value_per_pop = median_house_value / population
df_filtered['value_per_pop'] = df_filtered['median_house_value'] / df_filtered['population']

# 4. value_per_pop의 평균을 구한다. (소수 셋째자리까지 반올림)
result = round(df_filtered['value_per_pop'].mean(), 3)

# 결과
print("① iqr 값:", int(iqr))
print("② value_per_pop 평균값:", result)
```
