제공된 데이터 employee.csv는 IT 기업 직원의 나이, 연봉, 성별에 관한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Company: 근무 중인 회사명
2. Age: 나이
3. Salary: 연봉 (천 단위)
4. Gender: 성별 (0=남성, 1=여성)

[수행 순서]
1. 여성(1)의 인원 수를 A라고 한다.
2. Salary 값이 결측인 직원은 값을 0으로 대체한다.
3. 각 직원의 Gender 값과 Age에 따라 '대상' 여부를 판단하여 is_target컬럼을 추가한다.
※ 조건 : 여성(1)이고 나이가 30세 이상이면 'Yes', 아니면 'No'
4. is_target이 'Yes'인 직원들의 Salary 평균의 로그 값을 B라고 한다.

[문제]
1. 수행 순서 1)과 4)에서 구한 A, B의 값의 합(A+B)을 입력하시오.
☞ [제출 형식] 반올림하여 소수 셋째자리까지 작성

```python
import pandas as pd

df = pd.read_csv('../data/employee.csv')

# A = df[df['Gender'] == 1].shape[0]
A = df['Gender'].value_counts()[1]
print("A: ", A)
df['Salary'].fillna(0, inplace = True)

def isTarget(row):
    if (row['Gender'] == 1) & (row['Age'] >= 30):
        return 'Yes'
    else:
        return 'No'
    
df['is_target'] = df.apply(isTarget, axis=1)

# log 연산: numpy에서 임포트하기
import numpy as np
B = np.log(df[df['is_target'] == 'Yes']['Salary'].mean())
print("B: ", B)

print("ans: ", round(A+B, 3))

# 여기에 코드를 작성하시오.
```

# 풀이

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/1_DataSet/employee.csv')

import numpy as np

# 1. 여성(1)의 인원 수를 A라고 한다.
gender_counts = df['Gender'].value_counts()
A = gender_counts[1]

# 2. Salary 값이 결측인 직원은 값을 0으로 대체한다.
df['Salary'] = df['Salary'].fillna(0)

# 3. 각 직원의 Gender 값과 Age에 따라 '대상' 여부를 판단하여 is_target컬럼을 추가한다.
# ※ 조건 : 여성(1)이고 나이가 30세 이상이면 'Yes', 아니면 'No'
def determine_target(row):
    if row['Gender'] == 1 and row['Age'] >= 30:
        return 'Yes'
    else:
        return 'No'

df['is_target'] = df.apply(determine_target, axis=1)

# 4. is_target이 'Yes'인 직원들의 Salary 평균의 로그 값을 B라고 한다.
yes_salary_mean = df[df['is_target'] == 'Yes']['Salary'].mean()
B = np.log(yes_salary_mean)

# 최종 결과: A + B (소수 셋째자리 반올림)
result = round(A + B, 3)
print(result)
```
