제공된 데이터 employee.csv는 IT 기업 직원의 나이, 연봉, 성별에 관한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Company: 근무 중인 회사명
2. Age: 나이
3. Salary: 연봉 (천 단위)
4. Gender: 성별 (0=남성, 1=여성)

[수행 순서]
1. Salary 값이 결측인 직원은 이전 행의 값으로 결측치를 채운다.
2. Salary 컬럼을 기준으로 Min-Max 정규화와 Z-score 표준화를 각각 수행한다.
3. 정규화된 연봉이 0.8 이상인 직원의 수를 구한다.
4. 표준화된 연봉이 1.0 이상인 직원들의 평균 연봉을 구한다.

[문제]
1. 수행 순서 3)에서 계산한 값을 입력하시오.
☞ [제출 형식] 정수(integer)로 작성
2. 수행 순서 4)에서 계산한 값을 입력하시오.
☞ [제출 형식] 반올림하여 소수 셋째자리까지 작성

```python
import pandas as pd

df = pd.read_csv('../data/employee.csv')
df['Salary'] = df['Salary'].ffill()

from sklearn.preprocessing import MinMaxScaler, StandardScaler

m_scaler = MinMaxScaler()
# 인자 2차원으로 넘겨주는 거 까먹지 말기
df['normalized_Salary'] = m_scaler.fit_transform(df[['Salary']])
ans1 = int(df[df['normalized_Salary'] >= 0.8].shape[0])
# ans1 = (df['normalized_Salary'] >= 0.8).sum() -> true인 것들의 개수 총합
print("ans1: ", ans1)

s_scaler = StandardScaler()
df['standardized_Salary'] = s_scaler.fit_transform(df[['Salary']])
ans2 = round(df['Salary'][df['standardized_Salary'] >= 1.0].mean(),3)
print("ans2: ", ans2)


# 여기에 코드를 작성하시오.
```

# 풀이

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/1_DataSet/employee.csv')

from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 1. Salary 값이 결측인 직원은 이전 행의 값으로 결측치를 채운다.
df['Salary'] = df['Salary'].ffill()

# 2. Salary 컬럼을 기준으로 Min-Max 정규화와 Z-score 표준화를 각각 수행한다.
minmax_scaler = MinMaxScaler()
df['Salary_minmax'] = minmax_scaler.fit_transform(df[['Salary']])
standard_scaler = StandardScaler()
df['Salary_zscore'] = standard_scaler.fit_transform(df[['Salary']])

#3. 정규화된 연봉이 0.8 이상인 직원의 수를 구한다.
count_minmax_over_08 = (df['Salary_minmax'] >= 0.8).sum()

# 4. 표준화된 연봉이 1.0 이상인 직원들의 평균 연봉을 구한다.
mean_salary_zscore_over_1 = df[df['Salary_zscore'] >= 1.0]['Salary'].mean()
mean_salary_zscore_over_1_rounded = round(mean_salary_zscore_over_1, 3)

# 결과 출력
print("① 정규화된 연봉이 0.8이상인 직원 수 : ", count_minmax_over_08)
print("② 표준화된 연봉이 1.0이상인 직원의 평균 연봉 :", mean_salary_zscore_over_1_rounded)
```
