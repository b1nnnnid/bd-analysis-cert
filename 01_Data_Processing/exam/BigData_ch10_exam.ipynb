제공된 데이터 biketrip.csv는 자전거 대여 서비스 이용 내역이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Start Time : 대여 시작 시각
2. Stop Time : 반납 시각
3. Station ID : 대여 시작 지점 ID
4. UserType : 이용자 유형

[수행 순서]
1. Start Time과 Stop Time의 차이를 분 단위로 계산하여 새로운 컬럼 TripDuration을 생성한다.
2. 가장 오래 대여한 사용자가 대여한 Station ID를 구한다.
3. UserType별로 평균 TripDuration을 구한다.
4. 평균 TripDuration이 가장 높은 UserType의 평균 시간(분)을 구한다.

[문제]
1. 수행 순서 2)에서 계산한 값을 입력하시오.
☞ [제출 형식] 정수(integer)로 작성
2. 수행 순서 3)에서 계산한 값을 입력하시오.
☞ [제출 형식] 버림하여 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv("../data/biketrip.csv")

df['Start Time'] = pd.to_datetime(df['Start Time'])
df['Stop Time'] = pd.to_datetime(df['Stop Time'], format='%m/%d/%Y %H:%M')
# 포맷 설정 까먹지 말기

# 분 단위 구할 땐 반드시 total_seconds 활용
df['TripDuration'] = (df['Stop Time'] - df['Start Time']).dt.total_seconds()/60
max_duration = df['TripDuration'].idxmax()

# 인덱스랑 칼럼명 활용하기
ans1 = int(df.loc[max_duration, 'Station ID'])
print("ans1: ", ans1)

mean_duration = df.groupby('UserType')['TripDuration'].mean()
ans2 = int(mean_duration.max())
print("ans2: ", ans2)

# 여기에 코드를 작성하시오.
```

# 풀이

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/BigData/1_DataSet/biketrip.csv")

# 1. Start Time과 Stop Time의 차이를 분 단위로 계산하여 새로운 컬럼 TripDuration을 생성한다.
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['Stop Time'] = pd.to_datetime(df['Stop Time'], format='%m/%d/%Y %H:%M')

df['TripDuration'] = (df['Stop Time'] - df['Start Time']).dt.total_seconds() / 60

# 2. 가장 오래 대여한 사용자가 대여한 Station ID를 구한다.
longest_station_id = int(df.loc[df['TripDuration'].idxmax(), 'Station ID'])

# 3. UserType별로 평균 TripDuration을 구한다.
avg_trip_by_user = df.groupby('UserType')['TripDuration'].mean()

# 4. 평균 TripDuration이 가장 높은 UserType의 평균 시간을 구한다.
max_avg_trip_duration = int(avg_trip_by_user.max())

# 결과
print("① 가장 오래 대여한 사용자가 대여한 Station ID :", longest_station_id)
print("② 평균 TripDuarion이 가장 높은 UserType의 평균 시간 :", max_avg_trip_duration)
```
