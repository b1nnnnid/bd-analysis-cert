제공된 데이터 employee.csv는 IT 기업 직원의 나이, 연봉, 성별에 관한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Company: 근무 중인 회사명
2. Age: 나이
3. Salary: 연봉 (천 단위)
4. Gender: 성별 (0=남성, 1=여성)

[수행 순서]
1. Salary 칼럼의 결측값은 해당 회사(Company)별 평균 Salary로 대체한다.
※ 평균값은 소수점 이하는 버림하여 정수로 채운다.
2. Company별 평균 연봉을 계산한다.
3. 평균 연봉이 가장 높은 회사의 평균 연봉을 구한다.

[문제]
1. 수행 순서 3)에서 계산한 값을 입력하시오.
☞ [제출 형식] 버림하여 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv('../data/employee.csv')

# 여기서 바로 int 적용 불가(그룹핑 된 시리즈에 바로 int() 변환 불가능)
df['Salary'] = df['Salary'].fillna(df.groupby('Company')['Salary'].transform('mean'))

# 그 다음 절차에서 astype()로 정수 변환 해주기
avg_sal = df.groupby('Company')['Salary'].mean().astype(int)
highest = avg_sal.idxmax()

# print(int(df[df['Company'] == highest]['Salary'].mean()))
print("ans:", int(avg_sal[highest]))

# 여기에 코드를 작성하시오.
```

# 풀이

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/1_DataSet/employee.csv')

# 1. Salary 칼럼의 결측값은 해당 회사(Company)별 평균 Salary로 대체한다.
# ※ 평균값은 소수점 이하는 버림하여 정수로 채운다.
df['Salary'] = df['Salary'].fillna(df.groupby('Company')['Salary'].transform('mean'))

# 2. Company별 평균 연봉을 계산한다.
company_salary_mean = df.groupby('Company')['Salary'].mean().astype(int)

# 3. 평균 연봉이 가장 높은 회사의 평균 연봉을 구한다.
max_salary_company = company_salary_mean.idxmax()
max_salary_value = int(company_salary_mean[max_salary_company])

# 결과 출력
print("① 평균 연봉이 가장 높은 회사의 평균 연봉 : ", max_salary_value)
```
