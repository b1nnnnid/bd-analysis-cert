제공된 데이터 employee.csv는 IT 기업 직원의 나이, 연봉, 성별에 관한 자료이다.
아래의 수행 순서에 따라 데이터를 처리하고, 구한 값을 소문항별로 [제출 형식]에 맞춰 입력하시오.

[제공 데이터]
1. Company: 근무 중인 회사명
2. Age: 나이
3. Salary: 연봉 (천 단위)
4. Gender: 성별 (0=남성, 1=여성)

[수행 순서]
1. Salary 값이 결측인 직원은 전체 Salary의 평균값으로 결측치를 채운다.
※ 평균은 소수점 이하는 버림한 정수로 채운다.
2. Age 값이 결측인 직원은 해당 행을 삭제한다.
3. Company가 'TCS'인 직원 중, 연봉(Salary)이 2500 이상인 직원 수를 구한다.
4. Gender가 여성(1)인 직원들의 연봉 평균을 구한다.

[문제]
1. 수행 순서 3)에서 계산한 값을 입력하시오.
☞ [제출 형식] 정수(integer)로 작성
2. 수행 순서 4)에서 계산한 값을 입력하시오.
☞ [제출 형식] 버림하여 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv('../data/employee.csv')

# print(df.describe())

df['Salary'] = df['Salary'].fillna(int(df['Salary'].mean()))

df = df.dropna(subset = ['Age'])

ans3 = df[(df['Company'] == 'TCS') & (df['Salary']>=2500)].shape[0] # shape으로 개수 구하기(행, 열)
ans4 = int(df[df['Gender'] ==1]['Salary'].mean())

print(ans3, ans4)

# 여기에 코드를 작성하시오.
```

# 풀이

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BigData/1_DataSet/employee.csv')

# 1. Salary 값이 결측인 직원은 전체 Salary의 평균값으로 결측치를 채운다.
# ※ 평균은 소수점 이하는 버림한 정수로 채운다.
salary_mean = int(df['Salary'].mean())
df['Salary'] = df['Salary'].fillna(salary_mean)

# 2. Age 값이 결측인 직원은 해당 행을 삭제한다.
df_cleaned = df.dropna(subset=['Age'])

# 3. Company가 'TCS'인 직원 중, 연봉(Salary)이 2500 이상인 직원 수를 구한다.
tcs_high_salary_count = df_cleaned[(df_cleaned['Company'] == 'TCS') & (df_cleaned['Salary'] >= 2500)].shape[0]

# 4. Gender가 여성(1)인 직원들의 연봉 평균을 구한다.
female_salary_mean = df_cleaned[df_cleaned['Gender'] == 1]['Salary'].mean()

# 결과 출력
print("① TCS 직원 중 연봉 2500 이상인 사람 수:", tcs_high_salary_count)
print("② 여성 직원들의 연봉 평균 :", int(female_salary_mean))
```
